```python
import os
import random
from glob import glob
from collections import defaultdict

import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import transforms
from torchvision.io import read_image
from torchvision.models import resnet18
import matplotlib.pyplot as plt
from tqdm import tqdm

# Kaggle install commands (if needed):
# !pip install tqdm scikit-learn torch torchvision

# ---------------------------
# Set random seeds for reproducibility.
# ---------------------------
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)

# ---------------------------
# Utility function to normalize super class names.
# ---------------------------
def normalize_superclass(s):
    return s.replace('-', ' ').lower().strip()

# ---------------------------
# Define the list of OOD super classes based on the research paper.
# ---------------------------
OOD_SUPER_CLASSES = set(normalize_superclass(x) for x in [
    "vehicles 1",
    "vehicles 2",
    "large man made outdoor things",
    "household electrical devices",
    "household furniture",
    "food containers",
])

# ---------------------------
# Modified Scan Function.
#
# Assumes structure: root/<super_class>/<sub_class>/*.png
# Returns a list of (img_path, super_class, fine_label) and sorted list of super classes.
# ---------------------------
def scan_dataset(root_dir):
    data = []
    label_map = {}
    sub_label_counter = 0
    super_classes = set()
    
    super_dirs = sorted(os.listdir(root_dir))
    for super_dir in super_dirs:
        super_path = os.path.join(root_dir, super_dir)
        if os.path.isdir(super_path):
            super_classes.add(super_dir)
            sub_dirs = sorted(os.listdir(super_path))
            for sub_dir in sub_dirs:
                sub_path = os.path.join(super_path, sub_dir)
                if os.path.isdir(sub_path):
                    if sub_dir not in label_map:
                        label_map[sub_dir] = sub_label_counter
                        sub_label_counter += 1
                    img_files = sorted(glob(os.path.join(sub_path, '*')))
                    for file in img_files:
                        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                            data.append((file, super_dir, label_map[sub_dir]))
    print(label_map)
    return data, label_map, sorted(list(super_classes))

# ---------------------------
# Stratified split of data into train and validation (only for ID samples).
# ---------------------------
def stratified_split(data, val_ratio=0.1):
    paths = [item[0] for item in data]
    labels = [item[1] for item in data]
    sss = StratifiedShuffleSplit(n_splits=1, test_size=val_ratio, random_state=42)
    train_idx, val_idx = next(sss.split(paths, labels))
    train_data = [data[i] for i in train_idx]
    val_data   = [data[i] for i in val_idx]
    return train_data, val_data

# ---------------------------
# Mark a fraction of training samples (ID samples) as unlabeled (set label = -1).
# ---------------------------
def create_unlabeled_data(data, unlabeled_ratio):
    data_by_class = defaultdict(list)
    for idx, (path, label) in enumerate(data):
        data_by_class[label].append(idx)
    new_data = data.copy()
    for label, indices in data_by_class.items():
        n = len(indices)
        n_unlabeled = int(np.round(unlabeled_ratio * n))
        unlabeled_indices = random.sample(indices, n_unlabeled)
        for idx in unlabeled_indices:
            new_data[idx] = (new_data[idx][0], -1)
    return new_data

# ---------------------------
# Custom Dataset.
# Returns either a single augmented view or two augmented views (weak & strong).
# ---------------------------
class CIFAR100FolderDataset(Dataset):
    def __init__(self, file_list, transform=None, return_two_views=False, weak_transform=None, strong_transform=None):
        self.file_list = file_list
        self.transform = transform
        self.return_two_views = return_two_views
        self.weak_transform = weak_transform
        self.strong_transform = strong_transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, index):
        img_path, label = self.file_list[index]
        img = torchvision.io.read_image(img_path).float() / 255.0
        img = transforms.ToPILImage()(img)
        if self.return_two_views:
            img_weak = self.weak_transform(img) if self.weak_transform else img
            img_strong = self.strong_transform(img) if self.strong_transform else img
            return img_weak, img_strong, label
        else:
            img = self.transform(img) if self.transform else img
            return img, label

# ---------------------------
# Data Augmentations.
#
# Weak augmentation: random crop and horizontal flip.
# Strong augmentation: uses RandAugment + RandomErasing.
# ---------------------------
normalize = transforms.Normalize(mean=[0.5071, 0.4865, 0.4409],
                                 std=[0.2673, 0.2564, 0.2761])
input_size = 32

weak_transform = transforms.Compose([
    transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    normalize,
])

try:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandAugment(num_ops=2, magnitude=9),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])
except Exception:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])

val_transform = transforms.Compose([
    transforms.Resize(input_size + 4),
    transforms.CenterCrop(input_size),
    transforms.ToTensor(),
    normalize,
])

# ---------------------------
# SeFOSS Model Definition.
#
# The model includes:
#  - A backbone (ResNet-18) for feature extraction.
#  - A classifier head.
#  - A projection head used in self-supervision.
# ---------------------------
class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        backbone = resnet18(weights=None)  # using pretrained weights (set False to train from scratch)
        self.feature_extractor = nn.Sequential(*(list(backbone.children())[:-1]))
        self.feature_dim = backbone.fc.in_features
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)

    def forward_backbone(self, x):
        features = self.feature_extractor(x)
        features = features.view(x.size(0), -1)
        return features

    def forward(self, x):
        features = self.forward_backbone(x)
        logits = self.classifier(features)
        return logits

    def forward_features(self, x):
        return self.forward_backbone(x)

    def forward_projection(self, features):
        return self.projection(features)

# ---------------------------
# Loss Functions.
#
# free_energy: Implements Eq. (3) of the paper.
# cosine_similarity_loss: Implements the self-supervision loss of Eq. (2).
# ---------------------------
def free_energy(logits):
    return -torch.logsumexp(logits, dim=1)

def cosine_similarity_loss(f1, f2):
    f1_norm = F.normalize(f1, p=2, dim=1)
    f2_norm = F.normalize(f2, p=2, dim=1)
    loss = - (f1_norm * f2_norm).sum(dim=1).mean()
    return loss

# ---------------------------
# Training Hyperparameters.
# ---------------------------
num_classes = 70  # for ID classes only
lambda_p = 1.0      # pseudo-labeling loss weight (Section 3.2)
lambda_ss = 1.0     # self-supervision loss weight (Section 3.1)
lambda_e = 0.1      # energy regularization loss weight (Section 3.3)

tau_id = 0.5        # free-energy threshold for pseudo inlier selection
tau_ood = 1.0       # free-energy threshold for pseudo outlier selection
margin_ood = 1.5    # margin for the hinge loss

lr0 = 0.01
momentum = 0.9
weight_decay = 5e-4
num_epochs = 40
pretrain_epochs = 5

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---------------------------
# Build Evaluation Sets for Open-set AUROC.
#
# For training evaluation:
#    - ID training samples are assigned label 1.
#    - OOD samples are assigned label 0.
#
# For validation evaluation:
#    - ID validation samples are label 1.
#    - And we take a random sample from the OOD pool with label 0.
# ---------------------------
def build_unlabeled_eval_list(train_id_data, ood_data, val_id_data, ood_pool_size=None):
    # Build training evaluation list
    train_unlabeled_eval = []
    for sample in train_id_data:
        train_unlabeled_eval.append((sample[0], 1))
    for sample in ood_data:
        train_unlabeled_eval.append((sample[0], 0))
    
    # Build validation evaluation list
    val_unlabeled_eval = []
    for sample in val_id_data:
        val_unlabeled_eval.append((sample[0], 1))
    if ood_pool_size is None:
        ood_pool_size = len(val_id_data)
    val_ood_samples = random.sample(ood_data, min(len(ood_data), ood_pool_size))
    for sample in val_ood_samples:
        val_unlabeled_eval.append((sample[0], 0))
    return train_unlabeled_eval, val_unlabeled_eval
TestData = 0
# def main(labeled_ratio_given, total_epochs, batchSize, ID_Labeled_Size):
#     data_root = "/kaggle/input/cifar-100/cifar_100"
    
#     # --- Scan dataset ---
#     all_data, label_map, super_list = scan_dataset(data_root)
#     print(f"Found {len(all_data)} images in {len(super_list)} super classes.")
    
#     # --- Designate OOD super classes based on manual list ---
#     ood_super = set(s for s in super_list if normalize_superclass(s) in OOD_SUPER_CLASSES)
#     print(f"Designated the following super classes as OOD: {ood_super}")
    
#     # --- Partition the data ---
#     id_data = [ (file, fine_label) for (file, sclass, fine_label) in all_data if sclass not in ood_super ]
#     ood_data = [ (file, -1) for (file, sclass, fine_label) in all_data if sclass in ood_super ]
#     print(f"ID samples: {len(id_data)}; OOD samples: {len(ood_data)}")

#     ood_data, test_ood_data = stratified_split(ood_data, val_ratio = 0.1)
#     # --- Stratified split for ID samples ---
#     train_id_data, val_id_data = stratified_split(id_data, val_ratio=0.1)
#     print(f"ID train: {len(train_id_data)}, ID val: {len(val_id_data)}")
#     val_id_data, test_id_data = stratified_split(val_id_data, val_ratio=0.5)
#     TestData = test_id_data + test_ood_data
#     # Mark a fraction of ID training samples as unlabeled.
#     train_id_data = create_unlabeled_data(train_id_data, unlabeled_ratio= 1- labeled_ratio_given)
    
#     # --- Build training splits ---
#     train_labeled_data = [x for x in train_id_data if x[1] != -1]
#     train_unlabeled_data = train_id_data + ood_data
#     val_data = val_id_data
    
#     print(f"Labeled train samples (ID only): {len(train_labeled_data)}")
#     print(f"Out of Order train samples (OOD): {len(ood_data)}")
#     print(f"Unlabeled train samples (ID only): {len(train_unlabeled_data)-len(ood_data)}")
#     print(f"Total unlabeled train samples (ID unlabeled + OOD): {len(train_unlabeled_data)}")
    
#     # --- Build evaluation lists for open-set AUROC ---
#     train_unlabeled_eval_list, val_unlabeled_eval_list = build_unlabeled_eval_list(train_id_data, ood_data, val_id_data)
#     eval_transform = val_transform
#     train_unlabeled_eval_ds = CIFAR100FolderDataset(train_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
#     val_unlabeled_eval_ds = CIFAR100FolderDataset(val_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
#     test_unlabeled_eval_ds = CIFAR100FolderDataset(TestData, transform=eval_transform, return_two_views=False)
#     # --- Create training and validation datasets ---
#     labeled_dataset = CIFAR100FolderDataset(train_labeled_data, transform=weak_transform, return_two_views=False)
#     unlabeled_dataset = CIFAR100FolderDataset(train_unlabeled_data, return_two_views=True, weak_transform=weak_transform, strong_transform=strong_transform)
#     val_dataset = CIFAR100FolderDataset(val_data, transform=val_transform, return_two_views=False)
    
#     # --- Mini-Batch Composition ---
#     labeled_batch_size = ID_Labeled_Size    # Number of labeled examples per batch.
#     total_batch_size = batchSize      # Total examples (labeled + unlabeled)
#     unlabeled_batch_size = total_batch_size - labeled_batch_size
    
#     labeled_loader = DataLoader(labeled_dataset, batch_size=labeled_batch_size, shuffle=True, num_workers=4)
#     unlabeled_loader = DataLoader(unlabeled_dataset, batch_size=unlabeled_batch_size, shuffle=True, num_workers=4)
#     val_loader = DataLoader(val_dataset, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
#     train_unlabeled_eval_loader = DataLoader(train_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
#     val_unlabeled_eval_loader = DataLoader(val_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
#     labeled_iter = iter(labeled_loader)
#     unlabeled_iter = iter(unlabeled_loader)
    
#     model = SeFOSS(num_classes=num_classes, proj_dim=128).to(device)
#     optimizer = optim.SGD(model.parameters(), lr=lr0, momentum=momentum, weight_decay=weight_decay, nesterov=True)
#     scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_epochs)
def main(labeled_ratio_given, total_epochs, batchSize, ID_Labeled_Size):
    data_root = "/kaggle/input/cifar-100/cifar_100"
    
    # --- Scan dataset ---
    all_data, label_map, super_list = scan_dataset(data_root)
    print(f"Found {len(all_data)} images in {len(super_list)} super classes.")
    
    # --- Designate OOD super classes ---
    ood_super = set(s for s in super_list if normalize_superclass(s) in OOD_SUPER_CLASSES)
    print(f"Designated the following super classes as OOD: {ood_super}")
    
    # --- Identify fine-grained classes for OOD ---
    ood_fine_classes = set()
    for file, sclass, fine_label in all_data:
        if sclass in ood_super:
            ood_fine_classes.add(fine_label)
    
    ood_fine_classes = sorted(list(ood_fine_classes))
    print(f"OOD fine-grained classes (count: {len(ood_fine_classes)}): {ood_fine_classes}")
    
    # --- Verify OOD classes count ---
    if len(ood_fine_classes) != 30:
        raise ValueError(f"Expected 30 OOD classes, but found {len(ood_fine_classes)}. Adjust OOD_SUPER_CLASSES.")
    
    # --- Create new label map for ID classes ---
    id_fine_classes = sorted(set(fine_label for _, _, fine_label in all_data if fine_label not in ood_fine_classes))
    if len(id_fine_classes) != 70:
        raise ValueError(f"Expected 70 ID classes, but found {len(id_fine_classes)}.")
    
    # Map ID fine-grained labels to [0, 79]
    new_label_map = {old_label: new_label for new_label, old_label in enumerate(id_fine_classes)}
    print(f"Reassigned ID labels . Number of ID classes: {len(new_label_map)}")
    print(new_label_map)
    # --- Partition the data with new labels ---
    id_data = []
    ood_data = []
    for file, sclass, fine_label in all_data:
        if sclass in ood_super:
            ood_data.append((file, -1))
        else:
            if fine_label in new_label_map:
                id_data.append((file, new_label_map[fine_label]))
            else:
                raise ValueError(f"Unexpected fine_label {fine_label} not in new_label_map.")
    
    print(f"ID samples: {len(id_data)}; OOD samples: {len(ood_data)}")
    
    
    
    # --- Split OOD data for test set ---
    ood_data, test_ood_data = stratified_split(ood_data, val_ratio=0.1)
    
    # --- Stratified split for ID samples ---
    train_id_data, val_id_data = stratified_split(id_data, val_ratio=0.1)
    print(f"ID train: {len(train_id_data)}, ID val: {len(val_id_data)}")
    val_id_data, test_id_data = stratified_split(val_id_data, val_ratio=0.5)
    TestData = test_id_data + test_ood_data
    
    # --- Mark a fraction of ID training samples as unlabeled ---
    train_id_data = create_unlabeled_data(train_id_data, unlabeled_ratio=1 - labeled_ratio_given)
    
    # --- Build training splits ---
    train_labeled_data = [x for x in train_id_data if x[1] != -1]
    train_unlabeled_data = train_id_data + ood_data
    val_data = val_id_data
    
    print(f"Labeled train samples (ID only): {len(train_labeled_data)}")
    print(f"Out of Distribution train samples (OOD): {len(ood_data)}")
    print(f"Unlabeled train samples (ID only): {len(train_unlabeled_data) - len(ood_data)}")
    print(f"Total unlabeled train samples (ID unlabeled + OOD): {len(train_unlabeled_data)}")
    
    # --- Build evaluation lists for open-set AUROC ---
    train_unlabeled_eval_list, val_unlabeled_eval_list = build_unlabeled_eval_list(
        train_id_data, ood_data, val_id_data
    )
    eval_transform = val_transform
    train_unlabeled_eval_ds = CIFAR100FolderDataset(
        train_unlabeled_eval_list, transform=eval_transform, return_two_views=False
    )
    val_unlabeled_eval_ds = CIFAR100FolderDataset(
        val_unlabeled_eval_list, transform=eval_transform, return_two_views=False
    )
    test_unlabeled_eval_ds = CIFAR100FolderDataset(
        TestData, transform=eval_transform, return_two_views=False
    )
    
    # --- Create training and validation datasets ---
    labeled_dataset = CIFAR100FolderDataset(
        train_labeled_data, transform=weak_transform, return_two_views=False
    )
    unlabeled_dataset = CIFAR100FolderDataset(
        train_unlabeled_data,
        return_two_views=True,
        weak_transform=weak_transform,
        strong_transform=strong_transform,
    )
    val_dataset = CIFAR100FolderDataset(
        val_data, transform=val_transform, return_two_views=False
    )
    
    # --- Mini-Batch Composition ---
    labeled_batch_size = ID_Labeled_Size
    total_batch_size = batchSize
    unlabeled_batch_size = total_batch_size - labeled_batch_size
    
    labeled_loader = DataLoader(
        labeled_dataset, batch_size=labeled_batch_size, shuffle=True, num_workers=4
    )
    unlabeled_loader = DataLoader(
        unlabeled_dataset, batch_size=unlabeled_batch_size, shuffle=True, num_workers=4
    )
    val_loader = DataLoader(
        val_dataset, batch_size=total_batch_size, shuffle=False, num_workers=4
    )
    
    train_unlabeled_eval_loader = DataLoader(
        train_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4
    )
    val_unlabeled_eval_loader = DataLoader(
        val_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4
    )
    
    labeled_iter = iter(labeled_loader)
    unlabeled_iter = iter(unlabeled_loader)
    
    # --- Initialize model ---
    model = SeFOSS(num_classes=num_classes, proj_dim=128).to(device)
    optimizer = optim.SGD(
        model.parameters(),
        lr=lr0,
        momentum=momentum,
        weight_decay=weight_decay,
        nesterov=True,
    )
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_epochs)
    # Lists for plotting metrics per epoch.
    train_loss_list = []
    val_loss_list = []
    train_acc_list = []
    val_acc_list = []
    train_auroc_list = []
    val_auroc_list = []
    
    model.train()
    global_step = 0
    best_val_acc = 0.0
        # Hyperparameters for adaptive thresholds (Eq. 8)
    zeta_id = 0.2      # Controls τ_id (lower = more aggressive pseudo-labeling)
    zeta_ood = 1.3     # Controls τ_ood (higher = more conservative OOD detection)
    xi_ood = 1.9       # Controls margin_ood (higher = pushes OOD further away)
    
    # Initialize thresholds (will be updated after pre-training)
    tau_id = 0.0
    tau_ood = 0.0
    margin_ood = 0.0
    S_m = 0
    S_iqr = 0
    for epoch in range(total_epochs):
        if epoch == pretrain_epochs:
            print("Calculating adaptive thresholds...")
            model.eval()
            energy_scores = []
            with torch.no_grad():
                for imgs, labels in labeled_loader:
                    imgs = imgs.to(device)
                    logits = model(imgs)
                    fe = free_energy(logits)
                    energy_scores.extend(fe.cpu().numpy())
        
            energy_scores = np.array(energy_scores)
            S_m = np.median(energy_scores)
            S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        
            # Set thresholds according to Eq. 8
            tau_id = S_m - S_iqr * zeta_id
            tau_ood = S_m + S_iqr * zeta_ood
            margin_ood = S_m + S_iqr * xi_ood
        
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
            model.train()

        running_loss = 0.0
        num_batches = 0
        
        epoch_supervised_correct = 0
        epoch_supervised_total = 0
        
        pbar = tqdm(range(len(unlabeled_loader)), desc=f"Epoch [{epoch+1}/{total_epochs}]")
        for _ in pbar:
            try:
                labeled_images, labels = next(labeled_iter)
            except StopIteration:
                labeled_iter = iter(labeled_loader)
                labeled_images, labels = next(labeled_iter)
            try:
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            
            labeled_images = labeled_images.to(device)
            labels = labels.to(device)
            unlabeled_weak = unlabeled_weak.to(device)
            unlabeled_strong = unlabeled_strong.to(device)
            
            optimizer.zero_grad()
            
            # --- Supervised Loss & Accuracy (Eq. 10) ---
            logits_l = model(labeled_images)
            loss_supervised = F.cross_entropy(logits_l, labels)
            preds_l = torch.argmax(logits_l, dim=1)
            epoch_supervised_correct += (preds_l == labels).sum().item()
            epoch_supervised_total += labels.size(0)
            
            # --- Self-Supervision Loss (Eq. 2) ---
            features_weak = model.forward_features(unlabeled_weak)
            features_strong = model.forward_features(unlabeled_strong)
            proj_weak = model.forward_projection(features_weak)
            proj_strong = model.forward_projection(features_strong)
            loss_ss = cosine_similarity_loss(proj_strong, proj_weak.detach())
            
            # --- Pseudo-Labeling Loss (Eq. 4) ---
            logits_unlabeled_weak = model(unlabeled_weak)
            logits_unlabeled_strong = model(unlabeled_strong)
            with torch.no_grad():
                fe = free_energy(logits_unlabeled_weak)
                # Use fixed thresholds during pretraining, adaptive after
                current_tau_id = 0.5 if epoch < pretrain_epochs else tau_id
                current_tau_ood = 1.0 if epoch < pretrain_epochs else tau_ood
                current_margin_ood = 1.5 if epoch < pretrain_epochs else margin_ood
            
                pseudo_mask = (fe < current_tau_id)
                pseudo_labels = torch.argmax(F.softmax(logits_unlabeled_weak, dim=1), dim=1)

            if pseudo_mask.sum() > 0:
                loss_pseudo = F.cross_entropy(logits_unlabeled_strong[pseudo_mask], pseudo_labels[pseudo_mask])
            else:
                loss_pseudo = 0.0 * loss_supervised
            
            # --- Energy Regularization Loss (Eq. 7) ---
            with torch.no_grad():
                fe_for_energy = free_energy(logits_unlabeled_weak)
                energy_mask = (fe_for_energy > current_tau_ood)

            if energy_mask.sum() > 0:
                diff = (margin_ood - fe_for_energy[energy_mask]).clamp(min=0)
                loss_energy = diff.mean()
            else:
                loss_energy = 0.0 * loss_supervised
            
            # --- Total Loss (Eq. 9) ---
            if epoch < pretrain_epochs:
                loss_total = loss_supervised + lambda_ss * loss_ss
            else:
                loss_total = loss_supervised + lambda_ss * loss_ss + lambda_p * loss_pseudo + lambda_e * loss_energy
            
            loss_total.backward()
            optimizer.step()
            
            running_loss += loss_total.item()
            num_batches += 1
            global_step += 1
            pbar.set_postfix(loss=running_loss/num_batches)
        
        scheduler.step()
        avg_train_loss = running_loss / num_batches
        train_acc = (epoch_supervised_correct / epoch_supervised_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Training Loss: {avg_train_loss:.4f}, Training Accuracy: {train_acc:.2f}%")
        
        # --- Supervised Validation Evaluation (ID only) ---
        model.eval()
        val_loss = 0.0
        val_batches = 0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                labels = labels.to(device)
                outputs = model(imgs)
                loss_val = F.cross_entropy(outputs, labels)
                val_loss += loss_val.item()
                val_batches += 1
                preds = torch.argmax(outputs, dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)
        avg_val_loss = val_loss / val_batches
        val_acc = (val_correct / val_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_acc:.2f}%")
        
        # --- Open-set AUROC Evaluation ---
        # Training evaluation set:
        train_scores = []
        train_truth = []
        with torch.no_grad():
            for imgs, truth in train_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()  # Lower free-energy indicates ID
                train_scores.extend(scores)
                train_truth.extend(truth.numpy().tolist())
        try:
            train_auroc = roc_auc_score(train_truth, train_scores)
        except Exception as e:
            train_auroc = float('nan')
            print("Train AUROC computation error:", e)
        
        # Validation evaluation set:
        val_scores = []
        val_truth = []
        with torch.no_grad():
            for imgs, truth in val_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()
                val_scores.extend(scores)
                val_truth.extend(truth.numpy().tolist())
        try:
            val_auroc = roc_auc_score(val_truth, val_scores)
        except Exception as e:
            val_auroc = float('nan')
            print("Validation AUROC computation error:", e)
        
        print(f"Epoch [{epoch+1}/{total_epochs}] - Train AUROC: {train_auroc:.4f}, Validation AUROC: {val_auroc:.4f}")
        
        # Save best model based on validation accuracy.
        if val_acc > best_val_acc and epoch>pretrain_epochs:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f"best_sefoss_cifar100_{labeled_ratio_given*100:.2f}_percent_.pth")
            print(f"Best model updated (Val Acc: {best_val_acc:.2f}%)")
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
        
        # Log metrics for plotting.
        train_loss_list.append(avg_train_loss)
        val_loss_list.append(avg_val_loss)
        train_acc_list.append(train_acc)
        val_acc_list.append(val_acc)
        train_auroc_list.append(train_auroc)
        val_auroc_list.append(val_auroc)
        
        model.train()
    
    print("Training complete.")
    
    # ---------------------------
    # Plotting the metrics.
    # ---------------------------
    epochs = range(1, total_epochs + 1)
    
    # Plot Training and Validation Loss.
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_loss_list, label="Train Loss")
    plt.plot(epochs, val_loss_list, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Training and Validation Loss for Labeled to Unlabeled Ratio: {labeled_ratio_given*100}%, Batch Size: {batchSize}, ID_Labeled_per_Batch_Size:{ID_Labeled_Size}")
    plt.legend()
    plt.grid(True)
    plt.savefig("loss_plot.png")
    plt.show()
    
    # Plot Training and Validation Accuracy.
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_acc_list, label="Train Accuracy")
    plt.plot(epochs, val_acc_list, label="Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.savefig("accuracy_plot.png")
    plt.show()
    
    # Plot AUROC curves.
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_auroc_list, label="Train AUROC")
    plt.plot(epochs, val_auroc_list, label="Validation AUROC")
    plt.xlabel("Epoch")
    plt.ylabel("AUROC")
    plt.title("Open-set AUROC")
    plt.legend()
    plt.grid(True)
    plt.savefig("auroc_plot.png")
    plt.show()

# main(labeled_ratio_given = 0.10, total_epochs = 20, batchSize = 128, ID_Labeled_Size = 64)
# main(labeled_ratio_given = 0.25, total_epochs = 20, batchSize = 128, ID_Labeled_Size = 64)
# main(labeled_ratio_given = 0.50, total_epochs = 30, batchSize = 128, ID_Labeled_Size = 64)
# main(labeled_ratio_given = 0.66, total_epochs = 20, batchSize = 128, ID_Labeled_Size = 64)


main(labeled_ratio_given = 0.250, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 64)
main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 128)
# main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 128, ID_Labeled_Size = 64)

# main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 32)
# main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 128)
# main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 160)
# main(labeled_ratio_given = 0.50, total_epochs = 40, batchSize = 256, ID_Labeled_Size = 192)



```

```python
import os
import random
from glob import glob
from collections import defaultdict

import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import transforms
from torchvision.io import read_image
import matplotlib.pyplot as plt
from tqdm import tqdm

# Kaggle install commands (if needed):
# !pip install tqdm scikit-learn torch torchvision

def mish(x):
    return x * torch.tanh(F.softplus(x))

class PSBatchNorm2d(nn.BatchNorm2d):
    def __init__(self, num_features, alpha=0.1, eps=1e-05, momentum=0.001, affine=True, track_running_stats=True):
        super().__init__(num_features, eps, momentum, affine, track_running_stats)
        self.alpha = alpha

    def forward(self, x):
        return super().forward(x) + self.alpha

class BasicBlock(nn.Module):
    def __init__(self, in_planes, out_planes, stride, drop_rate=0.0, activate_before_residual=False):
        super(BasicBlock, self).__init__()
        self.bn1 = nn.BatchNorm2d(in_planes, momentum=0.001)
        self.relu1 = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.conv1 = nn.Conv2d(in_planes, out_planes, kernel_size=3, stride=stride,
                               padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_planes, momentum=0.001)
        self.relu2 = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.conv2 = nn.Conv2d(out_planes, out_planes, kernel_size=3, stride=1,
                               padding=1, bias=False)
        self.drop_rate = drop_rate
        self.equalInOut = (in_planes == out_planes)
        self.convShortcut = (not self.equalInOut) and nn.Conv2d(in_planes, out_planes, kernel_size=1, stride=stride,
                                                                padding=0, bias=False) or None
        self.activate_before_residual = activate_before_residual

    def forward(self, x):
        if not self.equalInOut and self.activate_before_residual:
            x = self.relu1(self.bn1(x))
        else:
            out = self.relu1(self.bn1(x))
        out = self.relu2(self.bn2(self.conv1(out if self.equalInOut else x)))
        if self.drop_rate > 0:
            out = F.dropout(out, p=self.drop_rate, training=self.training)
        out = self.conv2(out)
        return torch.add(x if self.equalInOut else self.convShortcut(x), out)

class NetworkBlock(nn.Module):
    def __init__(self, nb_layers, in_planes, out_planes, block, stride, drop_rate=0.0, activate_before_residual=False):
        super(NetworkBlock, self).__init__()
        self.layer = self._make_layer(block, in_planes, out_planes, nb_layers, stride, drop_rate, activate_before_residual)

    def _make_layer(self, block, in_planes, out_planes, nb_layers, stride, drop_rate, activate_before_residual):
        layers = []
        for i in range(int(nb_layers)):
            layers.append(block(i == 0 and in_planes or out_planes, out_planes,
                                i == 0 and stride or 1, drop_rate, activate_before_residual))
        return nn.Sequential(*layers)

    def forward(self, x):
        return self.layer(x)

class WideResNet(nn.Module):
    def __init__(self, num_classes, depth=28, widen_factor=2, drop_rate=0.0):
        super(WideResNet, self).__init__()
        channels = [16, 16*widen_factor, 32*widen_factor, 64*widen_factor]
        assert((depth - 4) % 6 == 0)
        n = (depth - 4) / 6
        block = BasicBlock
        self.conv1 = nn.Conv2d(3, channels[0], kernel_size=3, stride=1,
                               padding=1, bias=False)
        self.block1 = NetworkBlock(n, channels[0], channels[1], block, 1, drop_rate, activate_before_residual=True)
        self.block2 = NetworkBlock(n, channels[1], channels[2], block, 2, drop_rate)
        self.block3 = NetworkBlock(n, channels[2], channels[3], block, 2, drop_rate)
        self.bn1 = nn.BatchNorm2d(channels[3], momentum=0.001)
        self.relu = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.feature_dim = channels[3]
        self.fc = nn.Linear(channels[3], num_classes)

        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='leaky_relu')
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1.0)
                nn.init.constant_(m.bias, 0.0)
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.constant_(m.bias, 0.0)

    def forward_features(self, x):
        out = self.conv1(x)
        out = self.block1(out)
        out = self.block2(out)
        out = self.block3(out)
        out = self.relu(self.bn1(out))
        out = F.adaptive_avg_pool2d(out, 1)
        out = out.view(-1, self.feature_dim)
        return out

    def forward(self, x):
        return self.forward_features(x)

# ---------------------------
# Set random seeds for reproducibility.
# ---------------------------
def set_seed(seed=469):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(469)

# ---------------------------
# Utility function to normalize super class names.
# ---------------------------
def normalize_superclass(s):
    return s.replace('-', ' ').lower().strip()

# ---------------------------
# Define the list of OOD super classes based on the research paper.
# ---------------------------
OOD_SUPER_CLASSES = set(normalize_superclass(x) for x in [
    "vehicles 1",
    "vehicles 2",
    "large man made outdoor things",
    "household electrical devices",
    "household furniture",
    "food containers",
])

# ---------------------------
# Modified Scan Function.
#
# Assumes structure: root/<super_class>/<sub_class>/*.png
# Returns a list of (img_path, super_class, fine_label) and sorted list of super classes.
# ---------------------------
def scan_dataset(root_dir):
    data = []
    label_map = {}
    sub_label_counter = 0
    super_classes = set()
    
    super_dirs = sorted(os.listdir(root_dir))
    for super_dir in super_dirs:
        super_path = os.path.join(root_dir, super_dir)
        if os.path.isdir(super_path):
            super_classes.add(super_dir)
            sub_dirs = sorted(os.listdir(super_path))
            for sub_dir in sub_dirs:
                sub_path = os.path.join(super_path, sub_dir)
                if os.path.isdir(sub_path):
                    if sub_dir not in label_map:
                        label_map[sub_dir] = sub_label_counter
                        sub_label_counter += 1
                    img_files = sorted(glob(os.path.join(sub_path, '*')))
                    for file in img_files:
                        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                            data.append((file, super_dir, label_map[sub_dir]))
    print(label_map)
    return data, label_map, sorted(list(super_classes))

# ---------------------------
# Stratified split of data into train and validation (only for ID samples).
# ---------------------------
def stratified_split(data, val_ratio=0.1):
    paths = [item[0] for item in data]
    labels = [item[1] for item in data]
    sss = StratifiedShuffleSplit(n_splits=1, test_size=val_ratio, random_state=42)
    train_idx, val_idx = next(sss.split(paths, labels))
    train_data = [data[i] for i in train_idx]
    val_data = [data[i] for i in val_idx]
    return train_data, val_data

# ---------------------------
# Mark a fraction of training samples (ID samples) as unlabeled (set label = -1).
# ---------------------------
def create_unlabeled_data(data, unlabeled_ratio):
    data_by_class = defaultdict(list)
    for idx, (path, label) in enumerate(data):
        data_by_class[label].append(idx)
    new_data = data.copy()
    for label, indices in data_by_class.items():
        n = len(indices)
        n_unlabeled = int(np.round(unlabeled_ratio * n))
        unlabeled_indices = random.sample(indices, n_unlabeled)
        for idx in unlabeled_indices:
            new_data[idx] = (new_data[idx][0], -1)
    return new_data

# ---------------------------
# Custom Dataset.
# Returns either a single augmented view or two augmented views (weak & strong).
# ---------------------------
class CIFAR100FolderDataset(Dataset):
    def __init__(self, file_list, transform=None, return_two_views=False, weak_transform=None, strong_transform=None):
        self.file_list = file_list
        self.transform = transform
        self.return_two_views = return_two_views
        self.weak_transform = weak_transform
        self.strong_transform = strong_transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, index):
        img_path, label = self.file_list[index]
        img = torchvision.io.read_image(img_path).float() / 255.0
        img = transforms.ToPILImage()(img)
        if self.return_two_views:
            img_weak = self.weak_transform(img) if self.weak_transform else img
            img_strong = self.strong_transform(img) if self.strong_transform else img
            return img_weak, img_strong, label
        else:
            img = self.transform(img) if self.transform else img
            return img, label

# ---------------------------
# Data Augmentations.
#
# Weak augmentation: random crop and horizontal flip.
# Strong augmentation: uses RandAugment + RandomErasing.
# ---------------------------
normalize = transforms.Normalize(mean=[0.5071, 0.4865, 0.4409],
                                 std=[0.2673, 0.2564, 0.2761])
input_size = 32

weak_transform = transforms.Compose([
    transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    normalize,
])

try:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandAugment(num_ops=2, magnitude=9),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])
except Exception:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])

val_transform = transforms.Compose([
    transforms.Resize(input_size + 4),
    transforms.CenterCrop(input_size),
    transforms.ToTensor(),
    normalize,
])

# ---------------------------
# SeFOSS Model Definition.
#
# The model includes:
#  - A backbone (WideResNet) for feature extraction.
#  - A classifier head.
#  - A projection head used in self-supervision.
# ---------------------------
class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        backbone = WideResNet(num_classes=num_classes, depth=28, widen_factor=2, drop_rate=0.0)
        self.feature_extractor = backbone
        self.feature_dim = backbone.feature_dim
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)

    def forward_backbone(self, x):
        features = self.feature_extractor.forward_features(x)
        return features

    def forward(self, x):
        features = self.forward_backbone(x)
        logits = self.classifier(features)
        return logits

    def forward_features(self, x):
        return self.forward_backbone(x)

    def forward_projection(self, features):
        return self.projection(features)

# ---------------------------
# Loss Functions.
#
# free_energy: Implements Eq. (3) of the paper.
# cosine_similarity_loss: Implements the self-supervision loss of Eq. (2).
# ---------------------------
def free_energy(logits):
    return -torch.logsumexp(logits, dim=1)

def cosine_similarity_loss(f1, f2):
    f1_norm = F.normalize(f1, p=2, dim=1)
    f2_norm = F.normalize(f2, p=2, dim=1)
    loss = - (f1_norm * f2_norm).sum(dim=1).mean()
    return loss

# ---------------------------
# Training Hyperparameters.
# ---------------------------
num_classes = 100  # Full CIFAR-100 dataset
lambda_p = 1.0      # pseudo-labeling loss weight (Section 3.2)
lambda_ss = 1.0     # self-supervision loss weight (Section 3.1)
lambda_e = 0.1      # energy regularization loss weight (Section 3.3)

tau_id = 0.5        # free-energy threshold for pseudo inlier selection
tau_ood = 1.0       # free-energy threshold for pseudo outlier selection
margin_ood = 1.5    # margin for the hinge loss

lr0 = 0.01
momentum = 0.9
weight_decay = 5e-4
num_epochs = 40
pretrain_epochs = 5

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---------------------------
# Build Evaluation Sets for Open-set AUROC.
#
# For training evaluation:
#    - ID training samples are assigned label 1.
#    - OOD samples are assigned label 0.
#
# For validation evaluation:
#    - ID validation samples are label 1.
#    - And we take a random sample from the OOD pool with label 0.
# ---------------------------
def build_unlabeled_eval_list(train_id_data, ood_data, val_id_data, ood_pool_size=None):
    train_unlabeled_eval = []
    for sample in train_id_data:
        train_unlabeled_eval.append((sample[0], 1))
    for sample in ood_data:
        train_unlabeled_eval.append((sample[0], 0))
    
    val_unlabeled_eval = []
    for sample in val_id_data:
        val_unlabeled_eval.append((sample[0], 1))
    if ood_pool_size is None:
        ood_pool_size = len(val_id_data)
    val_ood_samples = random.sample(ood_data, min(len(ood_data), ood_pool_size))
    for sample in val_ood_samples:
        val_unlabeled_eval.append((sample[0], 0))
    return train_unlabeled_eval, val_unlabeled_eval

def main(labeled_ratio_given, total_epochs, batchSize, ID_Labeled_Size, name):
    data_root = "/kaggle/input/cifar-100/cifar_100"
    
    # --- Scan dataset ---
    all_data, label_map, super_list = scan_dataset(data_root)
    print(f"Found {len(all_data)} images in {len(super_list)} super classes.")
    
    # --- Designate OOD super classes based on manual list ---
    ood_super = set(s for s in super_list if normalize_superclass(s) in OOD_SUPER_CLASSES)
    print(f"Designated the following super classes as OOD: {ood_super}")
    
    # --- Partition the data ---
    id_data = [(file, fine_label) for (file, sclass, fine_label) in all_data if sclass not in ood_super]
    ood_data = [(file, -1) for (file, sclass, fine_label) in all_data if sclass in ood_super]
    print(f"ID samples: {len(id_data)}; OOD samples: {len(ood_data)}")

    ood_data, test_ood_data = stratified_split(ood_data, val_ratio=0.1)
    # --- Stratified split for ID samples ---
    train_id_data, val_id_data = stratified_split(id_data, val_ratio=0.1)
    print(f"ID train: {len(train_id_data)}, ID val: {len(val_id_data)}")
    val_id_data, test_id_data = stratified_split(val_id_data, val_ratio=0.5)
    TestData = test_id_data + test_ood_data
    
    # Mark a fraction of ID training samples as unlabeled.
    train_id_data = create_unlabeled_data(train_id_data, unlabeled_ratio=1 - labeled_ratio_given)
    
    # --- Build training splits ---
    train_labeled_data = [x for x in train_id_data if x[1] != -1]
    train_unlabeled_data = train_id_data + ood_data
    val_data = val_id_data
    
    print(f"Labeled train samples (ID only): {len(train_labeled_data)}")
    print(f"Out of Order train samples (OOD): {len(ood_data)}")
    print(f"Unlabeled train samples (ID only): {len(train_unlabeled_data) - len(ood_data)}")
    print(f"Total unlabeled train samples (ID unlabeled + OOD): {len(train_unlabeled_data)}")
    
    # --- Build evaluation lists for open-set AUROC ---
    train_unlabeled_eval_list, val_unlabeled_eval_list = build_unlabeled_eval_list(train_id_data, ood_data, val_id_data)
    eval_transform = val_transform
    train_unlabeled_eval_ds = CIFAR100FolderDataset(train_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
    val_unlabeled_eval_ds = CIFAR100FolderDataset(val_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
    test_unlabeled_eval_ds = CIFAR100FolderDataset(TestData, transform=eval_transform, return_two_views=False)
    
    # --- Create training and validation datasets ---
    labeled_dataset = CIFAR100FolderDataset(train_labeled_data, transform=weak_transform, return_two_views=False)
    unlabeled_dataset = CIFAR100FolderDataset(train_unlabeled_data, return_two_views=True, weak_transform=weak_transform, strong_transform=strong_transform)
    val_dataset = CIFAR100FolderDataset(val_data, transform=val_transform, return_two_views=False)
    
    # --- Mini-Batch Composition ---
    labeled_batch_size = ID_Labeled_Size
    total_batch_size = batchSize
    unlabeled_batch_size = total_batch_size - labeled_batch_size
    
    labeled_loader = DataLoader(labeled_dataset, batch_size=labeled_batch_size, shuffle=True, num_workers=4)
    unlabeled_loader = DataLoader(unlabeled_dataset, batch_size=unlabeled_batch_size, shuffle=True, num_workers=4)
    val_loader = DataLoader(val_dataset, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
    train_unlabeled_eval_loader = DataLoader(train_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
    val_unlabeled_eval_loader = DataLoader(val_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
    labeled_iter = iter(labeled_loader)
    unlabeled_iter = iter(unlabeled_loader)
    
    model = SeFOSS(num_classes=num_classes, proj_dim=128).to(device)
    optimizer = optim.SGD(model.parameters(), lr=lr0, momentum=momentum, weight_decay=weight_decay, nesterov=True)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_epochs)
    
    # Lists for plotting metrics per epoch.
    train_loss_list = []
    val_loss_list = []
    train_acc_list = []
    val_acc_list = []
    train_auroc_list = []
    val_auroc_list = []
    
    model.train()
    global_step = 0
    best_val_acc = 0.0
    
    # Hyperparameters for adaptive thresholds (Eq. 8)
    zeta_id = 0.2      # Controls τ_id (lower = more aggressive pseudo-labeling)
    zeta_ood = 1.3     # Controls τ_ood (higher = more conservative OOD detection)
    xi_ood = 1.9       # Controls margin_ood (higher = pushes OOD further away)
    
    # Initialize thresholds (will be updated after pre-training)
    tau_id = 0.0
    tau_ood = 0.0
    margin_ood = 0.0
    S_m = 0
    S_iqr = 0
    
    for epoch in range(total_epochs):
        if epoch == pretrain_epochs:
            print("Calculating adaptive thresholds...")
            model.eval()
            energy_scores = []
            with torch.no_grad():
                for imgs, labels in labeled_loader:
                    imgs = imgs.to(device)
                    logits = model(imgs)
                    fe = free_energy(logits)
                    energy_scores.extend(fe.cpu().numpy())
        
            energy_scores = np.array(energy_scores)
            S_m = np.median(energy_scores)
            S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        
            # Set thresholds according to Eq. 8
            tau_id = S_m - S_iqr * zeta_id
            tau_ood = S_m + S_iqr * zeta_ood
            margin_ood = S_m + S_iqr * xi_ood
        
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
            model.train()

        running_loss = 0.0
        num_batches = 0
        
        epoch_supervised_correct = 0
        epoch_supervised_total = 0
        
        pbar = tqdm(range(len(unlabeled_loader)), desc=f"Epoch [{epoch+1}/{total_epochs}]")
        for _ in pbar:
            try:
                labeled_images, labels = next(labeled_iter)
            except StopIteration:
                labeled_iter = iter(labeled_loader)
                labeled_images, labels = next(labeled_iter)
            try:
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            
            labeled_images = labeled_images.to(device)
            labels = labels.to(device)
            unlabeled_weak = unlabeled_weak.to(device)
            unlabeled_strong = unlabeled_strong.to(device)
            
            optimizer.zero_grad()
            
            # --- Supervised Loss & Accuracy (Eq. 10) ---
            logits_l = model(labeled_images)
            loss_supervised = F.cross_entropy(logits_l, labels)
            preds_l = torch.argmax(logits_l, dim=1)
            epoch_supervised_correct += (preds_l == labels).sum().item()
            epoch_supervised_total += labels.size(0)
            
            # --- Self-Supervision Loss (Eq. 2) ---
            features_weak = model.forward_features(unlabeled_weak)
            features_strong = model.forward_features(unlabeled_strong)
            proj_weak = model.forward_projection(features_weak)
            proj_strong = model.forward_projection(features_strong)
            loss_ss = cosine_similarity_loss(proj_strong, proj_weak.detach())
            
            # --- Pseudo-Labeling Loss (Eq. 4) ---
            logits_unlabeled_weak = model(unlabeled_weak)
            logits_unlabeled_strong = model(unlabeled_strong)
            with torch.no_grad():
                fe = free_energy(logits_unlabeled_weak)
                # Use fixed thresholds during pretraining, adaptive after
                current_tau_id = 0.5 if epoch < pretrain_epochs else tau_id
                current_tau_ood = 1.0 if epoch < pretrain_epochs else tau_ood
                current_margin_ood = 1.5 if epoch < pretrain_epochs else margin_ood
            
                pseudo_mask = (fe < current_tau_id)
                pseudo_labels = torch.argmax(F.softmax(logits_unlabeled_weak, dim=1), dim=1)

            if pseudo_mask.sum() > 0:
                loss_pseudo = F.cross_entropy(logits_unlabeled_strong[pseudo_mask], pseudo_labels[pseudo_mask])
            else:
                loss_pseudo = 0.0 * loss_supervised
            
            # --- Energy Regularization Loss (Eq. 7) ---
            with torch.no_grad():
                fe_for_energy = free_energy(logits_unlabeled_weak)
                energy_mask = (fe_for_energy > current_tau_ood)

            if energy_mask.sum() > 0:
                diff = (margin_ood - fe_for_energy[energy_mask]).clamp(min=0)
                loss_energy = diff.mean()
            else:
                loss_energy = 0.0 * loss_supervised
            
            # --- Total Loss (Eq. 9) ---
            if epoch < pretrain_epochs:
                loss_total = loss_supervised + lambda_ss * loss_ss
            else:
                loss_total = loss_supervised + lambda_ss * loss_ss + lambda_p * loss_pseudo + lambda_e * loss_energy
            
            loss_total.backward()
            optimizer.step()
            
            running_loss += loss_total.item()
            num_batches += 1
            global_step += 1
            pbar.set_postfix(loss=running_loss/num_batches)
        
        scheduler.step()
        avg_train_loss = running_loss / num_batches
        train_acc = (epoch_supervised_correct / epoch_supervised_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Training Loss: {avg_train_loss:.4f}, Training Accuracy: {train_acc:.2f}%")
        
        # --- Supervised Validation Evaluation (ID only) ---
        model.eval()
        val_loss = 0.0
        val_batches = 0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                labels = labels.to(device)
                outputs = model(imgs)
                loss_val = F.cross_entropy(outputs, labels)
                val_loss += loss_val.item()
                val_batches += 1
                preds = torch.argmax(outputs, dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)
        avg_val_loss = val_loss / val_batches
        val_acc = (val_correct / val_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_acc:.2f}%")
        
        # --- Open-set AUROC Evaluation ---
        train_scores = []
        train_truth = []
        with torch.no_grad():
            for imgs, truth in train_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()  # Lower free-energy indicates ID
                train_scores.extend(scores)
                train_truth.extend(truth.numpy().tolist())
        try:
            train_auroc = roc_auc_score(train_truth, train_scores)
        except Exception as e:
            train_auroc = float('nan')
            print("Train AUROC computation error:", e)
        
        val_scores = []
        val_truth = []
        with torch.no_grad():
            for imgs, truth in val_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()
                val_scores.extend(scores)
                val_truth.extend(truth.numpy().tolist())
        try:
            val_auroc = roc_auc_score(val_truth, val_scores)
        except Exception as e:
            val_auroc = float('nan')
            print("Validation AUROC computation error:", e)
        
        print(f"Epoch [{epoch+1}/{total_epochs}] - Train AUROC: {train_auroc:.4f}, Validation AUROC: {val_auroc:.4f}")
        
        # Save best model based on validation accuracy.
        if val_acc > best_val_acc and epoch > pretrain_epochs:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f"best_sefoss_cifar100_wideresnet_{name}_{epoch}_{labeled_ratio_given*100:.2f}_percent_.pth")
            print(f"Best model updated (Val Acc: {best_val_acc:.2f}%)")
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
        
        # Log metrics for plotting.
        train_loss_list.append(avg_train_loss)
        val_loss_list.append(avg_val_loss)
        train_acc_list.append(train_acc)
        val_acc_list.append(val_acc)
        train_auroc_list.append(train_auroc)
        val_auroc_list.append(val_auroc)
        
        model.train()
    
    print("Training complete.")
    
    # ---------------------------
    # Plotting the metrics.
    # ---------------------------
    epochs = range(1, total_epochs + 1)
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_loss_list, label="Train Loss")
    plt.plot(epochs, val_loss_list, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Training and Validation Loss for Labeled to Unlabeled Ratio: {labeled_ratio_given*100}%, Batch Size: {batchSize}, ID_Labeled_per_Batch_Size:{ID_Labeled_Size}")
    plt.legend()
    plt.grid(True)
    plt.savefig("loss_plot.png")
    plt.show()
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_acc_list, label="Train Accuracy")
    plt.plot(epochs, val_acc_list, label="Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.savefig("accuracy_plot.png")
    plt.show()
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_auroc_list, label="Train AUROC")
    plt.plot(epochs, val_auroc_list, label="Validation AUROC")
    plt.xlabel("Epoch")
    plt.ylabel("AUROC")
    plt.title("Open-set AUROC")
    plt.legend()
    plt.grid(True)
    plt.savefig("auroc_plot.png")
    plt.show()
```

```python
import os
import random
from glob import glob
from collections import defaultdict
from sklearn.metrics import roc_auc_score, classification_report

import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import transforms
from torchvision.io import read_image
import matplotlib.pyplot as plt
from tqdm import tqdm

# Kaggle install commands (if needed):
# !pip install tqdm scikit-learn torch torchvision

def mish(x):
    return x * torch.tanh(F.softplus(x))

class PSBatchNorm2d(nn.BatchNorm2d):
    def __init__(self, num_features, alpha=0.1, eps=1e-05, momentum=0.001, affine=True, track_running_stats=True):
        super().__init__(num_features, eps, momentum, affine, track_running_stats)
        self.alpha = alpha

    def forward(self, x):
        return super().forward(x) + self.alpha

class BasicBlock(nn.Module):
    def __init__(self, in_planes, out_planes, stride, drop_rate=0.0, activate_before_residual=False):
        super(BasicBlock, self).__init__()
        self.bn1 = nn.BatchNorm2d(in_planes, momentum=0.001)
        self.relu1 = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.conv1 = nn.Conv2d(in_planes, out_planes, kernel_size=3, stride=stride,
                               padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_planes, momentum=0.001)
        self.relu2 = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.conv2 = nn.Conv2d(out_planes, out_planes, kernel_size=3, stride=1,
                               padding=1, bias=False)
        self.drop_rate = drop_rate
        self.equalInOut = (in_planes == out_planes)
        self.convShortcut = (not self.equalInOut) and nn.Conv2d(in_planes, out_planes, kernel_size=1, stride=stride,
                                                                padding=0, bias=False) or None
        self.activate_before_residual = activate_before_residual

    def forward(self, x):
        if not self.equalInOut and self.activate_before_residual:
            x = self.relu1(self.bn1(x))
        else:
            out = self.relu1(self.bn1(x))
        out = self.relu2(self.bn2(self.conv1(out if self.equalInOut else x)))
        if self.drop_rate > 0:
            out = F.dropout(out, p=self.drop_rate, training=self.training)
        out = self.conv2(out)
        return torch.add(x if self.equalInOut else self.convShortcut(x), out)

class NetworkBlock(nn.Module):
    def __init__(self, nb_layers, in_planes, out_planes, block, stride, drop_rate=0.0, activate_before_residual=False):
        super(NetworkBlock, self).__init__()
        self.layer = self._make_layer(block, in_planes, out_planes, nb_layers, stride, drop_rate, activate_before_residual)

    def _make_layer(self, block, in_planes, out_planes, nb_layers, stride, drop_rate, activate_before_residual):
        layers = []
        for i in range(int(nb_layers)):
            layers.append(block(i == 0 and in_planes or out_planes, out_planes,
                                i == 0 and stride or 1, drop_rate, activate_before_residual))
        return nn.Sequential(*layers)

    def forward(self, x):
        return self.layer(x)

class WideResNet(nn.Module):
    def __init__(self, num_classes, depth=28, widen_factor=2, drop_rate=0.0):
        super(WideResNet, self).__init__()
        channels = [16, 16*widen_factor, 32*widen_factor, 64*widen_factor]
        assert((depth - 4) % 6 == 0)
        n = (depth - 4) / 6
        block = BasicBlock
        self.conv1 = nn.Conv2d(3, channels[0], kernel_size=3, stride=1,
                               padding=1, bias=False)
        self.block1 = NetworkBlock(n, channels[0], channels[1], block, 1, drop_rate, activate_before_residual=True)
        self.block2 = NetworkBlock(n, channels[1], channels[2], block, 2, drop_rate)
        self.block3 = NetworkBlock(n, channels[2], channels[3], block, 2, drop_rate)
        self.bn1 = nn.BatchNorm2d(channels[3], momentum=0.001)
        self.relu = nn.LeakyReLU(negative_slope=0.1, inplace=True)
        self.feature_dim = channels[3]
        self.fc = nn.Linear(channels[3], num_classes)

        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='leaky_relu')
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1.0)
                nn.init.constant_(m.bias, 0.0)
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.constant_(m.bias, 0.0)

    def forward_features(self, x):
        out = self.conv1(x)
        out = self.block1(out)
        out = self.block2(out)
        out = self.block3(out)
        out = self.relu(self.bn1(out))
        out = F.adaptive_avg_pool2d(out, 1)
        out = out.view(-1, self.feature_dim)
        return out

    def forward(self, x):
        return self.forward_features(x)

# ---------------------------
# Set random seeds for reproducibility.
# ---------------------------
def set_seed(seed=469):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(469)

# ---------------------------
# Utility function to normalize super class names.
# ---------------------------
def normalize_superclass(s):
    return s.replace('-', ' ').lower().strip()

# ---------------------------
# Define the list of OOD super classes based on the research paper.
# ---------------------------
OOD_SUPER_CLASSES = set(normalize_superclass(x) for x in [
    "vehicles 1",
    "vehicles 2",
    "large man made outdoor things",
    "household electrical devices",
    "household furniture",
    "food containers",
])

# ---------------------------
# Modified Scan Function.
#
# Assumes structure: root/<super_class>/<sub_class>/*.png
# Returns a list of (img_path, super_class, fine_label) and sorted list of super classes.
# ---------------------------
def scan_dataset(root_dir):
    data = []
    label_map = {}
    sub_label_counter = 0
    super_classes = set()
    
    super_dirs = sorted(os.listdir(root_dir))
    for super_dir in super_dirs:
        super_path = os.path.join(root_dir, super_dir)
        if os.path.isdir(super_path):
            super_classes.add(super_dir)
            sub_dirs = sorted(os.listdir(super_path))
            for sub_dir in sub_dirs:
                sub_path = os.path.join(super_path, sub_dir)
                if os.path.isdir(sub_path):
                    if sub_dir not in label_map:
                        label_map[sub_dir] = sub_label_counter
                        sub_label_counter += 1
                    img_files = sorted(glob(os.path.join(sub_path, '*')))
                    for file in img_files:
                        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                            data.append((file, super_dir, label_map[sub_dir]))
    print(label_map)
    return data, label_map, sorted(list(super_classes))

# ---------------------------
# Stratified split of data into train and validation (only for ID samples).
# ---------------------------
def stratified_split(data, val_ratio=0.1):
    paths = [item[0] for item in data]
    labels = [item[1] for item in data]
    sss = StratifiedShuffleSplit(n_splits=1, test_size=val_ratio, random_state=42)
    train_idx, val_idx = next(sss.split(paths, labels))
    train_data = [data[i] for i in train_idx]
    val_data = [data[i] for i in val_idx]
    return train_data, val_data

# ---------------------------
# Mark a fraction of training samples (ID samples) as unlabeled (set label = -1).
# ---------------------------
def create_unlabeled_data(data, unlabeled_ratio):
    data_by_class = defaultdict(list)
    for idx, (path, label) in enumerate(data):
        data_by_class[label].append(idx)
    new_data = data.copy()
    for label, indices in data_by_class.items():
        n = len(indices)
        n_unlabeled = int(np.round(unlabeled_ratio * n))
        unlabeled_indices = random.sample(indices, n_unlabeled)
        for idx in unlabeled_indices:
            new_data[idx] = (new_data[idx][0], -1)
    return new_data

# ---------------------------
# Custom Dataset.
# Returns either a single augmented view or two augmented views (weak & strong).
# ---------------------------
class CIFAR100FolderDataset(Dataset):
    def __init__(self, file_list, transform=None, return_two_views=False, weak_transform=None, strong_transform=None):
        self.file_list = file_list
        self.transform = transform
        self.return_two_views = return_two_views
        self.weak_transform = weak_transform
        self.strong_transform = strong_transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, index):
        img_path, label = self.file_list[index]
        img = torchvision.io.read_image(img_path).float() / 255.0
        img = transforms.ToPILImage()(img)
        if self.return_two_views:
            img_weak = self.weak_transform(img) if self.weak_transform else img
            img_strong = self.strong_transform(img) if self.strong_transform else img
            return img_weak, img_strong, label
        else:
            img = self.transform(img) if self.transform else img
            return img, label

# ---------------------------
# Data Augmentations.
#
# Weak augmentation: random crop and horizontal flip.
# Strong augmentation: uses RandAugment + RandomErasing.
# ---------------------------
normalize = transforms.Normalize(mean=[0.5071, 0.4865, 0.4409],
                                 std=[0.2673, 0.2564, 0.2761])
input_size = 32

weak_transform = transforms.Compose([
    transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    normalize,
])

try:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandAugment(num_ops=2, magnitude=9),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])
except Exception:
    strong_transform = transforms.Compose([
        transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
        transforms.ToTensor(),
        normalize,
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.33)),
    ])

val_transform = transforms.Compose([
    transforms.Resize(input_size + 4),
    transforms.CenterCrop(input_size),
    transforms.ToTensor(),
    normalize,
])

# ---------------------------
# SeFOSS Model Definition.
#
# The model includes:
#  - A backbone (WideResNet) for feature extraction.
#  - A classifier head.
#  - A projection head used in self-supervision.
# ---------------------------
class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        backbone = WideResNet(num_classes=num_classes, depth=28, widen_factor=2, drop_rate=0.0)
        self.feature_extractor = backbone
        self.feature_dim = backbone.feature_dim
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)

    def forward_backbone(self, x):
        features = self.feature_extractor.forward_features(x)
        return features

    def forward(self, x):
        features = self.forward_backbone(x)
        logits = self.classifier(features)
        return logits

    def forward_features(self, x):
        return self.forward_backbone(x)

    def forward_projection(self, features):
        return self.projection(features)

# ---------------------------
# Loss Functions.
#
# free_energy: Implements Eq. (3) of the paper.
# cosine_similarity_loss: Implements the self-supervision loss of Eq. (2).
# ---------------------------
def free_energy(logits):
    return -torch.logsumexp(logits, dim=1)

def cosine_similarity_loss(f1, f2):
    f1_norm = F.normalize(f1, p=2, dim=1)
    f2_norm = F.normalize(f2, p=2, dim=1)
    loss = - (f1_norm * f2_norm).sum(dim=1).mean()
    return loss

# ---------------------------
# Training Hyperparameters.
# ---------------------------
num_classes = 100  # Full CIFAR-100 dataset
lambda_p = 1.0      # pseudo-labeling loss weight (Section 3.2)
lambda_ss = 1.0     # self-supervision loss weight (Section 3.1)
lambda_e = 0.1      # energy regularization loss weight (Section 3.3)

tau_id = 0.5        # free-energy threshold for pseudo inlier selection
tau_ood = 1.0       # free-energy threshold for pseudo outlier selection
margin_ood = 1.5    # margin for the hinge loss

lr0 = 0.01
momentum = 0.9
weight_decay = 5e-4
num_epochs = 40
pretrain_epochs = 5

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---------------------------
# Build Evaluation Sets for Open-set AUROC.
#
# For training evaluation:
#    - ID training samples are assigned label 1.
#    - OOD samples are assigned label 0.
#
# For validation evaluation:
#    - ID validation samples are label 1.
#    - And we take a random sample from the OOD pool with label 0.
# ---------------------------
def build_unlabeled_eval_list(train_id_data, ood_data, val_id_data, ood_pool_size=None):
    train_unlabeled_eval = []
    for sample in train_id_data:
        train_unlabeled_eval.append((sample[0], 1))
    for sample in ood_data:
        train_unlabeled_eval.append((sample[0], 0))
    
    val_unlabeled_eval = []
    for sample in val_id_data:
        val_unlabeled_eval.append((sample[0], 1))
    if ood_pool_size is None:
        ood_pool_size = len(val_id_data)
    val_ood_samples = random.sample(ood_data, min(len(ood_data), ood_pool_size))
    for sample in val_ood_samples:
        val_unlabeled_eval.append((sample[0], 0))
    return train_unlabeled_eval, val_unlabeled_eval

def main(labeled_ratio_given, total_epochs, batchSize, ID_Labeled_Size, name):
    data_root = "/kaggle/input/cifar-100/cifar_100"
    
    # --- Scan dataset ---
    all_data, label_map, super_list = scan_dataset(data_root)
    print(f"Found {len(all_data)} images in {len(super_list)} super classes.")
    
    # --- Designate OOD super classes based on manual list ---
    ood_super = set(s for s in super_list if normalize_superclass(s) in OOD_SUPER_CLASSES)
    print(f"Designated the following super classes as OOD: {ood_super}")
    
    # --- Partition the data ---
    id_data = [(file, fine_label) for (file, sclass, fine_label) in all_data if sclass not in ood_super]
    ood_data = [(file, -1) for (file, sclass, fine_label) in all_data if sclass in ood_super]
    print(f"ID samples: {len(id_data)}; OOD samples: {len(ood_data)}")

    ood_data, test_ood_data = stratified_split(ood_data, val_ratio=0.1)
    # --- Stratified split for ID samples ---
    train_id_data, val_id_data = stratified_split(id_data, val_ratio=0.1)
    print(f"ID train: {len(train_id_data)}, ID val: {len(val_id_data)}")
    val_id_data, test_id_data = stratified_split(val_id_data, val_ratio=0.5)
    TestData = test_id_data + test_ood_data
    
    # Mark a fraction of ID training samples as unlabeled.
    train_id_data = create_unlabeled_data(train_id_data, unlabeled_ratio=1 - labeled_ratio_given)
    
    # --- Build training splits ---
    train_labeled_data = [x for x in train_id_data if x[1] != -1]
    train_unlabeled_data = train_id_data + ood_data
    val_data = val_id_data
    
    print(f"Labeled train samples (ID only): {len(train_labeled_data)}")
    print(f"Out of Order train samples (OOD): {len(ood_data)}")
    print(f"Unlabeled train samples (ID only): {len(train_unlabeled_data) - len(ood_data)}")
    print(f"Total unlabeled train samples (ID unlabeled + OOD): {len(train_unlabeled_data)}")
    
    # --- Build evaluation lists for open-set AUROC ---
    train_unlabeled_eval_list, val_unlabeled_eval_list = build_unlabeled_eval_list(train_id_data, ood_data, val_id_data)
    eval_transform = val_transform
    train_unlabeled_eval_ds = CIFAR100FolderDataset(train_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
    val_unlabeled_eval_ds = CIFAR100FolderDataset(val_unlabeled_eval_list, transform=eval_transform, return_two_views=False)
    test_unlabeled_eval_ds = CIFAR100FolderDataset(TestData, transform=eval_transform, return_two_views=False)
    
    # --- Create training and validation datasets ---
    labeled_dataset = CIFAR100FolderDataset(train_labeled_data, transform=weak_transform, return_two_views=False)
    unlabeled_dataset = CIFAR100FolderDataset(train_unlabeled_data, return_two_views=True, weak_transform=weak_transform, strong_transform=strong_transform)
    val_dataset = CIFAR100FolderDataset(val_data, transform=val_transform, return_two_views=False)
    
    # --- Mini-Batch Composition ---
    labeled_batch_size = ID_Labeled_Size
    total_batch_size = batchSize
    unlabeled_batch_size = total_batch_size - labeled_batch_size
    
    labeled_loader = DataLoader(labeled_dataset, batch_size=labeled_batch_size, shuffle=True, num_workers=4)
    unlabeled_loader = DataLoader(unlabeled_dataset, batch_size=unlabeled_batch_size, shuffle=True, num_workers=4)
    val_loader = DataLoader(val_dataset, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
    train_unlabeled_eval_loader = DataLoader(train_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
    val_unlabeled_eval_loader = DataLoader(val_unlabeled_eval_ds, batch_size=total_batch_size, shuffle=False, num_workers=4)
    
    labeled_iter = iter(labeled_loader)
    unlabeled_iter = iter(unlabeled_loader)
    
    model = SeFOSS(num_classes=num_classes, proj_dim=128).to(device)
    optimizer = optim.SGD(model.parameters(), lr=lr0, momentum=momentum, weight_decay=weight_decay, nesterov=True)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_epochs)
    
    # Lists for plotting metrics per epoch.
    train_loss_list = []
    val_loss_list = []
    train_acc_list = []
    val_acc_list = []
    train_auroc_list = []
    val_auroc_list = []
    
    model.train()
    global_step = 0
    best_val_acc = 0.0
    
    # Hyperparameters for adaptive thresholds (Eq. 8)
    zeta_id = 0.2      # Controls τ_id (lower = more aggressive pseudo-labeling)
    zeta_ood = 1.3     # Controls τ_ood (higher = more conservative OOD detection)
    xi_ood = 1.9       # Controls margin_ood (higher = pushes OOD further away)
    
    # Initialize thresholds (will be updated after pre-training)
    tau_id = 0.0
    tau_ood = 0.0
    margin_ood = 0.0
    S_m = 0
    S_iqr = 0
    
    for epoch in range(total_epochs):
        if epoch == pretrain_epochs:
            print("Calculating adaptive thresholds...")
            model.eval()
            energy_scores = []
            with torch.no_grad():
                for imgs, labels in labeled_loader:
                    imgs = imgs.to(device)
                    logits = model(imgs)
                    fe = free_energy(logits)
                    energy_scores.extend(fe.cpu().numpy())
        
            energy_scores = np.array(energy_scores)
            S_m = np.median(energy_scores)
            S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        
            # Set thresholds according to Eq. 8
            tau_id = S_m - S_iqr * zeta_id
            tau_ood = S_m + S_iqr * zeta_ood
            margin_ood = S_m + S_iqr * xi_ood
        
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
            model.train()

        running_loss = 0.0
        num_batches = 0
        
        epoch_supervised_correct = 0
        epoch_supervised_total = 0
        
        pbar = tqdm(range(len(unlabeled_loader)), desc=f"Epoch [{epoch+1}/{total_epochs}]")
        for _ in pbar:
            try:
                labeled_images, labels = next(labeled_iter)
            except StopIteration:
                labeled_iter = iter(labeled_loader)
                labeled_images, labels = next(labeled_iter)
            try:
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            except StopIteration:
                unlabeled_iter = iter(unlabeled_loader)
                unlabeled_weak, unlabeled_strong, _ = next(unlabeled_iter)
            
            labeled_images = labeled_images.to(device)
            labels = labels.to(device)
            unlabeled_weak = unlabeled_weak.to(device)
            unlabeled_strong = unlabeled_strong.to(device)
            
            optimizer.zero_grad()
            
            # --- Supervised Loss & Accuracy (Eq. 10) ---
            logits_l = model(labeled_images)
            loss_supervised = F.cross_entropy(logits_l, labels)
            preds_l = torch.argmax(logits_l, dim=1)
            epoch_supervised_correct += (preds_l == labels).sum().item()
            epoch_supervised_total += labels.size(0)
            
            # --- Self-Supervision Loss (Eq. 2) ---
            features_weak = model.forward_features(unlabeled_weak)
            features_strong = model.forward_features(unlabeled_strong)
            proj_weak = model.forward_projection(features_weak)
            proj_strong = model.forward_projection(features_strong)
            loss_ss = cosine_similarity_loss(proj_strong, proj_weak.detach())
            
            # --- Pseudo-Labeling Loss (Eq. 4) ---
            logits_unlabeled_weak = model(unlabeled_weak)
            logits_unlabeled_strong = model(unlabeled_strong)
            with torch.no_grad():
                fe = free_energy(logits_unlabeled_weak)
                # Use fixed thresholds during pretraining, adaptive after
                current_tau_id = 0.5 if epoch < pretrain_epochs else tau_id
                current_tau_ood = 1.0 if epoch < pretrain_epochs else tau_ood
                current_margin_ood = 1.5 if epoch < pretrain_epochs else margin_ood
            
                pseudo_mask = (fe < current_tau_id)
                pseudo_labels = torch.argmax(F.softmax(logits_unlabeled_weak, dim=1), dim=1)

            if pseudo_mask.sum() > 0:
                loss_pseudo = F.cross_entropy(logits_unlabeled_strong[pseudo_mask], pseudo_labels[pseudo_mask])
            else:
                loss_pseudo = 0.0 * loss_supervised
            
            # --- Energy Regularization Loss (Eq. 7) ---
            with torch.no_grad():
                fe_for_energy = free_energy(logits_unlabeled_weak)
                energy_mask = (fe_for_energy > current_tau_ood)

            if energy_mask.sum() > 0:
                diff = (margin_ood - fe_for_energy[energy_mask]).clamp(min=0)
                loss_energy = diff.mean()
            else:
                loss_energy = 0.0 * loss_supervised
            
            # --- Total Loss (Eq. 9) ---
            if epoch < pretrain_epochs:
                loss_total = loss_supervised + lambda_ss * loss_ss
            else:
                loss_total = loss_supervised + lambda_ss * loss_ss + lambda_p * loss_pseudo + lambda_e * loss_energy
            
            loss_total.backward()
            optimizer.step()
            
            running_loss += loss_total.item()
            num_batches += 1
            global_step += 1
            pbar.set_postfix(loss=running_loss/num_batches)
        
        scheduler.step()
        avg_train_loss = running_loss / num_batches
        train_acc = (epoch_supervised_correct / epoch_supervised_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Training Loss: {avg_train_loss:.4f}, Training Accuracy: {train_acc:.2f}%")
        
        # --- Supervised Validation Evaluation (ID only) ---
        model.eval()
        val_loss = 0.0
        val_batches = 0
        val_correct = 0
        val_total = 0
        val_labels_all = []
        val_preds_all = []
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                labels = labels.to(device)
                outputs = model(imgs)
                loss_val = F.cross_entropy(outputs, labels)
                val_loss += loss_val.item()
                val_batches += 1
                preds = torch.argmax(outputs, dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)
                # Collect labels and predictions
                val_labels_all.extend(labels.cpu().numpy())
                val_preds_all.extend(preds.cpu().numpy())
        avg_val_loss = val_loss / val_batches
        val_acc = (val_correct / val_total) * 100
        print(f"Epoch [{epoch+1}/{total_epochs}] - Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_acc:.2f}%")
        
        # --- Open-set AUROC Evaluation ---
        train_scores = []
        train_truth = []
        with torch.no_grad():
            for imgs, truth in train_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()  # Lower free-energy indicates ID
                train_scores.extend(scores)
                train_truth.extend(truth.numpy().tolist())
        try:
            train_auroc = roc_auc_score(train_truth, train_scores)
        except Exception as e:
            train_auroc = float('nan')
            print("Train AUROC computation error:", e)
        
        val_scores = []
        val_truth = []
        with torch.no_grad():
            for imgs, truth in val_unlabeled_eval_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                fe_score = free_energy(outputs)
                scores = (-fe_score.cpu().numpy()).tolist()
                val_scores.extend(scores)
                val_truth.extend(truth.numpy().tolist())
        try:
            val_auroc = roc_auc_score(val_truth, val_scores)
        except Exception as e:
            val_auroc = float('nan')
            print("Validation AUROC computation error:", e)
        
        print(f"Epoch [{epoch+1}/{total_epochs}] - Train AUROC: {train_auroc:.4f}, Validation AUROC: {val_auroc:.4f}")
        
        # --- Validation Classification Report ---
        print(f"Validation Classification Report (Epoch {epoch+1}):")
        try:
            print(classification_report(val_labels_all, val_preds_all, digits=4, zero_division=0))
        except Exception as e:
            print("Error generating classification report:", e)
        
        # Save best model based on validation accuracy.
        if val_acc > best_val_acc and epoch > pretrain_epochs:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f"best_sefoss_cifar100_wideresnet_{name}_{epoch}_{labeled_ratio_given*100:.2f}_percent_.pth")
            print(f"Best model updated (Val Acc: {best_val_acc:.2f}%)")
            print(f"Adaptive thresholds calculated:")
            print(f"  Median energy (S_m): {S_m:.4f}")
            print(f"  IQR (S_iqr): {S_iqr:.4f}")
            print(f"  τ_id: {tau_id:.4f}")
            print(f"  τ_ood: {tau_ood:.4f}")
            print(f"  margin_ood: {margin_ood:.4f}")
        
        # Log metrics for plotting.
        train_loss_list.append(avg_train_loss)
        val_loss_list.append(avg_val_loss)
        train_acc_list.append(train_acc)
        val_acc_list.append(val_acc)
        train_auroc_list.append(train_auroc)
        val_auroc_list.append(val_auroc)
        
        model.train()
    
    print("Training complete.")

    # --- Plotting the metrics ---
    epochs = range(1, total_epochs + 1)
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_loss_list, label="Train Loss")
    plt.plot(epochs, val_loss_list, label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Training and Validation Loss for Labeled to Unlabeled Ratio: {labeled_ratio_given*100}%, Batch Size: {batchSize}, ID_Labeled_per_Batch_Size:{ID_Labeled_Size}")
    plt.legend()
    plt.grid(True)
    plt.savefig("loss_plot.png")
    plt.show()
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_acc_list, label="Train Accuracy")
    plt.plot(epochs, val_acc_list, label="Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.savefig("accuracy_plot.png")
    plt.show()
    
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, train_auroc_list, label="Train AUROC")
    plt.plot(epochs, val_auroc_list, label="Validation AUROC")
    plt.xlabel("Epoch")
    plt.ylabel("AUROC")
    plt.title("Open-set AUROC")
    plt.legend()
    plt.grid(True)
    plt.savefig("auroc_plot.png")
    plt.show()

# Run the main function
```
