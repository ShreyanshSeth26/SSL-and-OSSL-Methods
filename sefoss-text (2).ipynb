```python
import os
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torch.amp import autocast, GradScaler
from transformers import AutoModel, AutoTokenizer, get_cosine_schedule_with_warmup
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, average_precision_score
from tqdm import tqdm  # Import tqdm for progress bars
import nltk
from nltk.corpus import wordnet
import warnings
from copy import deepcopy

# Suppress CUDA warnings
warnings.filterwarnings("ignore", message="Unable to register cuFFT factory")
warnings.filterwarnings("ignore", message="Unable to register cuDNN factory")
warnings.filterwarnings("ignore", message="Unable to register cuBLAS factory")

# Set random seeds
RandomState = 2022469
np.random.seed(RandomState)
random.seed(RandomState)
torch.manual_seed(RandomState)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True

def LoadPubmedRctFile(FilePath):
    # Load PubMed RCT text file into cases
    Cases = []
    with open(FilePath, "r", encoding="utf-8") as Fobj:
        Case = {"label": None, "text": ""}  # Current case
        for Line in Fobj:
            Line = Line.strip()
            if Line == "":
                # End of case
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Case = {"label": None, "text": ""}
            elif Line.split("\t")[0].isupper():
                # New labeled case
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Parts = Line.split("\t")
                Case = {"label": Parts[0], "text": Parts[1]}
            else:
                # Continuation text
                Case["text"] += " " + Line.strip()
        if Case["label"] and Case["text"]:
            # Final case
            Cases.append(Case)
    DF = pd.DataFrame(Cases)
    print(f"Dataset loaded with {len(DF)} samples.")
    print(DF.head())
    return DF

class LoadPubmedRct:
    def __init__(self, FilePath):
        self.DataFrame = LoadPubmedRctFile(FilePath)  # Load dataset
    def GetDataFrame(self):
        return self.DataFrame

class SplitEightyTenTen:
    def __init__(self, DataFrame):
        self.DataFrame = DataFrame
    def Split(self):
        # Split: 80% train, 10% val, 10% test
        TrainData, TempData = train_test_split(
            self.DataFrame,
            test_size=0.2,
            stratify=self.DataFrame["label"],
            random_state=RandomState
        )
        ValData, TestData = train_test_split(
            TempData,
            test_size=0.5,
            stratify=TempData["label"],
            random_state=RandomState
        )
        print("Train:", len(TrainData), "Val:", len(ValData), "Test:", len(TestData))
        return TrainData, ValData, TestData

class AugSynonymReplace:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        # Replace words with synonyms
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                Syns = wordnet.synsets(W)
                if Syns:
                    Lemmas = Syns[0].lemmas()
                    if Lemmas:
                        Synonym = Lemmas[0].name()
                        if Synonym != W:
                            NewWords.append(Synonym)
                            continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomMask:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
    def __call__(self, Text):
        # Mask one token
        Tokens = self.Tokenizer.tokenize(Text)
        if len(Tokens) == 0:
            return Text
        Idx = random.randint(0, len(Tokens)-1)
        Tokens[Idx] = self.Tokenizer.mask_token
        return self.Tokenizer.convert_tokens_to_string(Tokens)

class AugRandomDelete:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        # Delete words
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomSwap:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        # Swap two words
        Words = Text.split()
        if len(Words) < 2:
            return Text
        if random.random() < self.Probability:
            i = random.randint(0, len(Words)-1)
            j = random.randint(0, len(Words)-1)
            while j == i:
                j = random.randint(0, len(Words)-1)
            Words[i], Words[j] = Words[j], Words[i]
        return " ".join(Words)

class WeakAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.1)
        self.Mask = AugRandomMask(Tokenizer)
    def __call__(self, Text):
        # Weak augmentation
        SynText = self.SynReplace(Text)
        MaskedText = self.Mask(SynText)
        return MaskedText

class StrongAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.2)
        self.Delete = AugRandomDelete(Probability=0.15)
        self.Swap = AugRandomSwap(Probability=0.15)
    def __call__(self, Text):
        # Strong augmentation
        SynText = self.SynReplace(Text)
        DeletedText = self.Delete(SynText)
        SwappedText = self.Swap(DeletedText)
        return SwappedText

class PubmedRctSeFOSSDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories  # ID categories
        self.MaxLen = MaxLen
        self.IsUnlabeled = IsUnlabeled
        self.LabelList = sorted(self.Df["label"].unique())  # All labels
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.WeakAug = WeakAugmentText(Tokenizer)
        self.StrongAug = StrongAugmentText(Tokenizer)
        self.IsId = self.Df["label"].isin(self.IdCategories).values  # Flag ID/OOD
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        IsIdSample = self.IsId[Idx]
        Label = self.Label2Idx[LabelStr] if IsIdSample or not self.IsUnlabeled else -1
        # Encode texts
        OriginalEnc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        WeakText = self.WeakAug(Text)
        WeakEnc = self.Tokenizer.encode_plus(
            WeakText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        StrongText = self.StrongAug(Text)
        StrongEnc = self.Tokenizer.encode_plus(
            StrongText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "OriginalInputIds": OriginalEnc["input_ids"].squeeze(0),
            "OriginalAttentionMask": OriginalEnc["attention_mask"].squeeze(0),
            "WeakInputIds": WeakEnc["input_ids"].squeeze(0),
            "WeakAttentionMask": WeakEnc["attention_mask"].squeeze(0),
            "StrongInputIds": StrongEnc["input_ids"].squeeze(0),
            "StrongAttentionMask": StrongEnc["attention_mask"].squeeze(0),
            "Label": Label,
            "IsId": IsIdSample
        }

class PubmedRctTestDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories
        self.MaxLen = MaxLen
        self.LabelList = sorted(set(IdCategories))  # ID labels only
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.Df = self.Df[self.Df["label"].isin(self.IdCategories)].reset_index(drop=True)  # Filter ID
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        Label = self.Label2Idx[LabelStr]
        Enc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "InputIds": Enc["input_ids"].squeeze(0),
            "AttentionMask": Enc["attention_mask"].squeeze(0),
            "Label": Label
        }

class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        self.backbone = AutoModel.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")  # PubMedBERT
        self.feature_dim = self.backbone.config.hidden_size  # 768
        self.classifier = nn.Linear(self.feature_dim, num_classes)  # Classifier
        self.projection = nn.Linear(self.feature_dim, proj_dim)  # Projection head
    def forward(self, input_ids, attention_mask):
        # Compute logits
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        features = outputs.last_hidden_state[:, 0]
        logits = self.classifier(features)
        return logits
    def forward_features(self, input_ids, attention_mask):
        # Extract features
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        return outputs.last_hidden_state[:, 0]
    def forward_projection(self, features):
        # Project features
        return self.projection(features)

class TrainSeFOSSText:
    def __init__(
        self,
        Model,
        LabeledLoader,
        UnlabeledLoader,
        ValLoader,
        TestLoader,
        LabelList,
        Epochs=10,
        PretrainEpochs=1,
        Lr=2e-5,
        ZetaId=0.2,
        ZetaOod=1.5,
        XiOod=2.1,
        Wp=1.0,
        Ws=5.0,
        We=1.0,
        Ww=1e-4,
        EmaMomentum=0.999,
        Device=DEVICE,
        RunName="Run"
    ):
        self.Model = Model.to(Device)
        self.LabeledLoader = LabeledLoader
        self.UnlabeledLoader = UnlabeledLoader
        self.ValLoader = ValLoader
        self.TestLoader = TestLoader
        self.LabelList = LabelList
        self.NumLabels = len(LabelList)
        self.Epochs = Epochs
        self.PretrainEpochs = PretrainEpochs
        self.Lr = Lr
        self.ZetaId = ZetaId
        self.ZetaOod = ZetaOod
        self.XiOod = XiOod
        self.Wp = Wp
        self.Ws = Ws
        self.We = We
        self.Ww = Ww
        self.EmaMomentum = EmaMomentum
        self.Device = Device
        self.RunName = RunName
        self.TauId = None
        self.TauOod = None
        self.MarginOod = None
        self.EmaModel = None
    def free_energy(self, logits):
        # Free-energy score (Eq. 3)
        return -torch.logsumexp(logits, dim=1)
    def update_ema(self):
        # Update EMA parameters
        for ema_param, param in zip(self.EmaModel.parameters(), self.Model.parameters()):
            ema_param.data = self.EmaMomentum * ema_param.data + (1 - self.EmaMomentum) * param.data
    def compute_adaptive_thresholds(self):
        # Compute thresholds using EMA model
        self.Model.eval()
        self.EmaModel.eval()
        energy_scores = []
        with torch.no_grad():
            # Iterate over labeled data with tqdm
            for batch in tqdm(self.LabeledLoader, desc="Computing Thresholds", leave=False):
                input_ids = batch["OriginalInputIds"].to(self.Device)
                attention_mask = batch["OriginalAttentionMask"].to(self.Device)
                logits = self.EmaModel(input_ids, attention_mask)
                fe = self.free_energy(logits)
                energy_scores.extend(fe.cpu().numpy())
        energy_scores = np.array(energy_scores)
        S_m = np.median(energy_scores)
        S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        self.TauId = S_m - S_iqr * self.ZetaId
        self.TauOod = S_m + S_iqr * self.ZetaOod
        self.MarginOod = S_m + S_iqr * self.XiOod
        print(f"Adaptive thresholds calculated:")
        print(f"  Median energy (S_m): {S_m:.4f}")
        print(f"  IQR (S_iqr): {S_iqr:.4f}")
        print(f"  τ_id: {self.TauId:.4f}")
        print(f"  τ_ood: {self.TauOod:.4f}")
        print(f"  margin_ood: {self.MarginOod:.4f}")
        print(f"  Threshold gap: τ_ood - τ_id = {self.TauOod - self.TauId:.4f}")
        print(f"  Margin gap: m_ood - τ_ood = {self.MarginOod - self.TauOod:.4f}")
        self.Model.train()
    def Run(self):
        Optimizer = optim.AdamW(self.Model.parameters(), lr=self.Lr)
        TotalSteps = self.Epochs * len(self.LabeledLoader)
        Scheduler = get_cosine_schedule_with_warmup(
            Optimizer, num_warmup_steps=0, num_training_steps=TotalSteps
        )
        scaler = GradScaler()
        TrainLossList = []
        TrainSupLossList = []
        TrainSelfSupLossList = []
        TrainPseudoLossList = []
        TrainEnergyLossList = []
        TrainAccList = []
        ValLossList = []
        ValAccList = []
        # Pre-training
        self.EmaModel = deepcopy(self.Model).to(self.Device)
        for Ep in range(self.PretrainEpochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            # Use tqdm for pre-training loop
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} Pretrain E{Ep+1}/{self.PretrainEpochs}",
                leave=False
            ):
                # Supervised loss
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)  # Eq. 10
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                # Self-supervised loss
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    WeakProjNorm = F.normalize(WeakProj, dim=1)
                    StrongProjNorm = F.normalize(StrongProj, dim=1)
                    LossSelfSup = -torch.mean(torch.sum(WeakProjNorm * StrongProjNorm, dim=1))  # Eq. 2
                Loss = LossSup + self.Ws * LossSelfSup
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                NumBatches += 1
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(0.0)
            TrainEnergyLossList.append(0.0)
            TrainAccList.append(TrainAcc)
            # Validation with tqdm
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValFeScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} Pretrain E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValFeScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Pretrain Epoch[{Ep+1}/{self.PretrainEpochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val FE mean: {np.mean(ValFeScores):.4f}, std: {np.std(ValFeScores):.4f}")
        # Compute thresholds
        self.compute_adaptive_thresholds()
        # Main training
        for Ep in range(self.PretrainEpochs, self.Epochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            PseudoLossTotal = 0.0
            EnergyLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            # Use tqdm for main training loop
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} E{Ep+1}/{self.Epochs}",
                leave=False
            ):
                # Supervised loss
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)  # Eq. 10
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                # Unlabeled losses
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                IsIdBatch = UnlabeledBatch["IsId"].to(self.Device)
                # Self-supervised loss
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    WeakProjNorm = F.normalize(WeakProj, dim=1)
                    StrongProjNorm = F.normalize(StrongProj, dim=1)
                    LossSelfSup = -torch.mean(torch.sum(WeakProjNorm * StrongProjNorm, dim=1))  # Eq. 2
                # Pseudo-labeling
                with torch.no_grad():
                    LogitsUWeak = self.Model(UWeakIds, UWeakMask)
                    FeWeak = self.free_energy(LogitsUWeak)
                    PseudoMask = (FeWeak < self.TauId).float() * IsIdBatch.float()
                    _, PseudoLabels = torch.max(F.softmax(LogitsUWeak, dim=1), dim=1)
                with autocast(device_type="cuda"):
                    LogitsUStrong = self.Model(UStrongIds, UStrongMask)
                    LossPseudo = F.cross_entropy(LogitsUStrong, PseudoLabels, reduction="none")
                    LossPseudo = (LossPseudo * PseudoMask).mean()  # Eq. 4
                    if PseudoMask.sum() == 0:
                        LossPseudo = torch.tensor(0.0, device=self.Device)
                # Energy regularization
                OutlierMask = (FeWeak > self.TauOod).float()
                if OutlierMask.sum() > 0:
                    LossEnergy = torch.mean(
                        OutlierMask * torch.relu(self.MarginOod - FeWeak) ** 2
                    ) / (OutlierMask.sum() + 1e-8)  # Eq. 7
                else:
                    LossEnergy = torch.tensor(0.0, device=self.Device)
                # Weight regularization
                WeightNorm = sum(p.norm().pow(2) for p in self.Model.parameters()) * 0.5  # Eq. 11
                Loss = (
                    LossSup
                    + self.Wp * LossPseudo
                    + self.Ws * LossSelfSup
                    + self.We * LossEnergy
                    + self.Ww * WeightNorm
                )
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                Scheduler.step()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                PseudoLossTotal += LossPseudo.item()
                EnergyLossTotal += LossEnergy.item()
                NumBatches += 1
                # print(f"Pseudo-inliers: {PseudoMask.sum().item()}/{PseudoMask.size(0)}")
                # print(f"Energy loss: {LossEnergy.item():.4f}")
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            AvgPseudoLoss = PseudoLossTotal / NumBatches
            AvgEnergyLoss = EnergyLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(AvgPseudoLoss)
            TrainEnergyLossList.append(AvgEnergyLoss)
            TrainAccList.append(TrainAcc)
            # Validation with tqdm
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValFeScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValFeScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Epoch[{Ep+1}/{self.Epochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f},")
            print(f"Pseudo={AvgPseudoLoss:.4f}, Energy={AvgEnergyLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val FE mean: {np.mean(ValFeScores):.4f}, std: {np.std(ValFeScores):.4f}")
        # Test evaluation with tqdm
        TestPreds = []
        TestLabelsList = []
        TestProbsList = []
        TestFeScores = []
        self.Model.eval()
        with torch.no_grad():
            for Batch in tqdm(
                self.TestLoader,
                desc=f"{self.RunName} Test",
                leave=False
            ):
                Ids_ = Batch["InputIds"].to(self.Device)
                Mask_ = Batch["AttentionMask"].to(self.Device)
                Labels_ = Batch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    Logits_ = self.Model(Ids_, Mask_)
                    Probs_ = F.softmax(Logits_, dim=1)
                    Fe_ = self.free_energy(Logits_)
                Preds_ = torch.argmax(Logits_, dim=1)
                TestPreds.extend(Preds_.cpu().numpy())
                TestLabelsList.extend(Labels_.cpu().numpy())
                TestProbsList.extend(Probs_.cpu().numpy())
                TestFeScores.extend(Fe_.cpu().numpy())
        TestAcc = accuracy_score(TestLabelsList, TestPreds)
        OneHotLabels = np.eye(self.NumLabels)[np.array(TestLabelsList)]
        RocAuc = roc_auc_score(OneHotLabels, np.array(TestProbsList), multi_class="ovr")
        AvgPrecision = average_precision_score(OneHotLabels, np.array(TestProbsList), average="macro")
        print(f"\n{self.RunName} Final Test Results:")
        print(f"Test Accuracy: {TestAcc*100:.2f}%")
        print(f"ROC AUC: {RocAuc:.4f}, Average Precision: {AvgPrecision:.4f}")
        print(f"Test FE mean: {np.mean(TestFeScores):.4f}, std: {np.std(TestFeScores):.4f}")
        print("Classification Report:")
        print(classification_report(TestLabelsList, TestPreds, target_names=self.LabelList, zero_division=0))
        return {
            "TrainLoss": TrainLossList,
            "TrainSupLoss": TrainSupLossList,
            "TrainSelfSupLoss": TrainSelfSupLossList,
            "TrainPseudoLoss": TrainPseudoLossList,
            "TrainEnergyLoss": TrainEnergyLossList,
            "TrainAcc": TrainAccList,
            "ValLoss": ValLossList,
            "ValAcc": ValAccList,
            "TestAcc": TestAcc,
            "RocAuc": RocAuc,
            "AvgPrecision": AvgPrecision
        }

def Main():
    print("Starting the training")
    nltk.download("wordnet")
    FilePath = r"/kaggle/input/pubmed-rct/pubmed_rct.txt"
    Loader = LoadPubmedRct(FilePath)
    DF = Loader.GetDataFrame()
    Splitter = SplitEightyTenTen(DF)
    TrainDF, ValDF, TestDF = Splitter.Split()
    Tokenizer = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
    IdCategories = ["METHODS", "RESULTS"]
    LabeledSizes = [500, 1000, 5000]
    print(f"ID Categories: {IdCategories}")
    ResultsDict = {}
    
    for Size in LabeledSizes:
        print(f"\nRunning SeFOSS With {Size} Labeled ID Samples")
        TrainIdDF = TrainDF[TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        TrainOodDF = TrainDF[~TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        if len(TrainIdDF) < Size:
            raise ValueError(f"Requested {Size} labeled samples, but only {len(TrainIdDF)} ID samples available.")
        LabeledSplit, UnlabeledIdSplit = train_test_split(
            TrainIdDF,
            train_size=Size,
            stratify=TrainIdDF["label"],
            random_state=RandomState
        )
        UnlabeledSplit = pd.concat([UnlabeledIdSplit, TrainOodDF]).reset_index(drop=True)
        print(f"Labeled ID: {len(LabeledSplit)}, Unlabeled: {len(UnlabeledSplit)} (ID={len(UnlabeledIdSplit)}, OOD={len(TrainOodDF)})")
        LabeledDataset = PubmedRctSeFOSSDataset(
            LabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False
        )
        UnlabeledDataset = PubmedRctSeFOSSDataset(
            UnlabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=True
        )
        ValData = PubmedRctTestDataset(ValDF, Tokenizer, IdCategories, MaxLen=256)
        TestData = PubmedRctTestDataset(TestDF, Tokenizer, IdCategories, MaxLen=256)
        LabeledLoader = DataLoader(LabeledDataset, batch_size=4, shuffle=True)
        UnlabeledLoader = DataLoader(UnlabeledDataset, batch_size=4, shuffle=True)
        ValLoader = DataLoader(ValData, batch_size=4, shuffle=False)
        TestLoader = DataLoader(TestData, batch_size=4, shuffle=False)
        Model = SeFOSS(num_classes=len(IdCategories), proj_dim=128)
        Trainer = TrainSeFOSSText(
            Model,
            LabeledLoader,
            UnlabeledLoader,
            ValLoader,
            TestLoader,
            IdCategories,
            Epochs=10,
            PretrainEpochs=1,
            Lr=2e-5,
            ZetaId=0.2,
            ZetaOod=1.5,
            XiOod=2.1,
            Wp=1.0,
            Ws=0.5,
            We=1.0,
            Ww=1e-4,
            EmaMomentum=0.999,
            Device=DEVICE,
            RunName=f"PubMedRCT_{Size}_Labeled"
        )
        RunResults = Trainer.Run()
        torch.save(Model.state_dict(), f"SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        print(f"Model weights saved to SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        EpochsRange = range(1, len(RunResults["TrainLoss"]) + 1)
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Total Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSupLoss"], label="Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSelfSupLoss"], label="Self-Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainPseudoLoss"], label="Pseudo-Labeling Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainEnergyLoss"], label="Energy Loss", marker="o")
        plt.title(f"Training Loss Curves ({Size} Labeled)")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainAcc"], label="Train Accuracy", marker="o")
        plt.title(f"Training Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Validation Loss", marker="o")
        plt.title(f"Validation Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, [v * 100 for v in RunResults["ValAcc"]], label="Val Accuracy (%)", marker="o")
        plt.title(f"Validation Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy (%)")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Val Loss", marker="o")
        plt.title(f"Train Vs. Val Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()

        # plt.figure(figsize=(8, 5))
        # plt.plot(EpochsRange, RunResults["TrainAuroc"], label="Train AUROC", marker="o")
        # plt.plot(EpochsRange, RunResults["ValAuroc"], label="Val AUROC", marker="o")
        # plt.title(f"Train vs. Val AUROC (Size={Size})")
        # plt.xlabel("Epoch")
        # plt.ylabel("AUROC")
        # plt.legend()
        # plt.tight_layout()
        # plt.savefig(f"auroc_{Size}.png")
        # plt.show()
        
        # ResultsDict[Size] = RunResults
        # print(f"--- Finished Plotting For {Size} Labeled Samples ---")
    BestCount = None
    BestAcc = 0.0
    for Size in LabeledSizes:
        Acc = ResultsDict[Size]["TestAcc"]
        if Acc > BestAcc:
            BestAcc = Acc
            BestCount = Size
    print(f"\nBest Run: {BestCount} Labeled Samples With Test Accuracy: {BestAcc * 100:.2f}%")

Main()
```

```python
import os
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torch.amp import autocast, GradScaler
from transformers import AutoModel, AutoTokenizer, get_cosine_schedule_with_warmup
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, average_precision_score
from tqdm import tqdm
import nltk
from nltk.corpus import wordnet
import warnings
from copy import deepcopy

# Suppress CUDA warnings
warnings.filterwarnings("ignore", message="Unable to register cuFFT factory")
warnings.filterwarnings("ignore", message="Unable to register cuDNN factory")
warnings.filterwarnings("ignore", message="Unable to register cuBLAS factory")

# Set random seeds
RandomState = 2022469
np.random.seed(RandomState)
random.seed(RandomState)
torch.manual_seed(RandomState)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True

def LoadPubmedRctFile(FilePath):
    Cases = []
    with open(FilePath, "r", encoding="utf-8") as Fobj:
        Case = {"label": None, "text": ""}
        for Line in Fobj:
            Line = Line.strip()
            if Line == "":
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Case = {"label": None, "text": ""}
            elif Line.split("\t")[0].isupper():
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Parts = Line.split("\t")
                Case = {"label": Parts[0], "text": Parts[1]}
            else:
                Case["text"] += " " + Line.strip()
        if Case["label"] and Case["text"]:
            Cases.append(Case)
    DF = pd.DataFrame(Cases)
    print(f"Dataset loaded with {len(DF)} samples.")
    print(DF.head())
    return DF

class LoadPubmedRct:
    def __init__(self, FilePath):
        self.DataFrame = LoadPubmedRctFile(FilePath)
    def GetDataFrame(self):
        return self.DataFrame

class SplitEightyTenTen:
    def __init__(self, DataFrame):
        self.DataFrame = DataFrame
    def Split(self):
        TrainData, TempData = train_test_split(
            self.DataFrame,
            test_size=0.2,
            stratify=self.DataFrame["label"],
            random_state=RandomState
        )
        ValData, TestData = train_test_split(
            TempData,
            test_size=0.5,
            stratify=TempData["label"],
            random_state=RandomState
        )
        print("Train:", len(TrainData), "Val:", len(ValData), "Test:", len(TestData))
        return TrainData, ValData, TestData

class AugSynonymReplace:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                Syns = wordnet.synsets(W)
                if Syns:
                    Lemmas = Syns[0].lemmas()
                    if Lemmas:
                        Synonym = Lemmas[0].name()
                        if Synonym != W:
                            NewWords.append(Synonym)
                            continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomMask:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
    def __call__(self, Text):
        Tokens = self.Tokenizer.tokenize(Text)
        if len(Tokens) == 0:
            return Text
        Idx = random.randint(0, len(Tokens)-1)
        Tokens[Idx] = self.Tokenizer.mask_token
        return self.Tokenizer.convert_tokens_to_string(Tokens)

class AugRandomDelete:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomSwap:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        if len(Words) < 2:
            return Text
        if random.random() < self.Probability:
            i = random.randint(0, len(Words)-1)
            j = random.randint(0, len(Words)-1)
            while j == i:
                j = random.randint(0, len(Words)-1)
            Words[i], Words[j] = Words[j], Words[i]
        return " ".join(Words)

class WeakAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.1)
        self.Mask = AugRandomMask(Tokenizer)
    def __call__(self, Text):
        SynText = self.SynReplace(Text)
        MaskedText = self.Mask(SynText)
        return MaskedText

class StrongAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.2)
        self.Delete = AugRandomDelete(Probability=0.15)
        self.Swap = AugRandomSwap(Probability=0.15)
    def __call__(self, Text):
        SynText = self.SynReplace(Text)
        DeletedText = self.Delete(SynText)
        SwappedText = self.Swap(DeletedText)
        return SwappedText

class PubmedRctSeFOSSDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories
        self.MaxLen = MaxLen
        self.IsUnlabeled = IsUnlabeled
        self.LabelList = sorted(self.Df["label"].unique())
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.WeakAug = WeakAugmentText(Tokenizer)
        self.StrongAug = StrongAugmentText(Tokenizer)
        self.IsId = self.Df["label"].isin(self.IdCategories).values
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        IsIdSample = self.IsId[Idx]
        Label = self.Label2Idx[LabelStr] if IsIdSample or not self.IsUnlabeled else -1
        OriginalEnc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        WeakText = self.WeakAug(Text)
        WeakEnc = self.Tokenizer.encode_plus(
            WeakText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        StrongText = self.StrongAug(Text)
        StrongEnc = self.Tokenizer.encode_plus(
            StrongText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "OriginalInputIds": OriginalEnc["input_ids"].squeeze(0),
            "OriginalAttentionMask": OriginalEnc["attention_mask"].squeeze(0),
            "WeakInputIds": WeakEnc["input_ids"].squeeze(0),
            "WeakAttentionMask": WeakEnc["attention_mask"].squeeze(0),
            "StrongInputIds": StrongEnc["input_ids"].squeeze(0),
            "StrongAttentionMask": StrongEnc["attention_mask"].squeeze(0),
            "Label": Label,
            "IsId": IsIdSample
        }

class PubmedRctTestDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories
        self.MaxLen = MaxLen
        self.LabelList = sorted(set(IdCategories))
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.Df = self.Df[self.Df["label"].isin(self.IdCategories)].reset_index(drop=True)
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        Label = self.Label2Idx[LabelStr]
        Enc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "InputIds": Enc["input_ids"].squeeze(0),
            "AttentionMask": Enc["attention_mask"].squeeze(0),
            "Label": Label
        }

class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        self.backbone = AutoModel.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
        self.feature_dim = self.backbone.config.hidden_size
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)
    def forward(self, input_ids, attention_mask):
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        features = outputs.last_hidden_state[:, 0]
        logits = self.classifier(features)
        return logits
    def forward_features(self, input_ids, attention_mask):
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        return outputs.last_hidden_state[:, 0]
    def forward_projection(self, features):
        return self.projection(features)

class TrainSeFOSSText:
    def __init__(
        self,
        Model,
        LabeledLoader,
        UnlabeledLoader,
        ValLoader,
        TestLoader,
        LabelList,
        Epochs=10,
        PretrainEpochs=1,
        Lr=2e-5,
        ZetaId=0.2,
        ZetaOod=1.5,
        XiOod=2.1,
        Wp=1.0,
        Ws=0.5,
        We=1.0,
        Ww=1e-4,
        EmaMomentum=0.999,
        Device=DEVICE,
        RunName="Run"
    ):
        self.Model = Model.to(Device)
        self.LabeledLoader = LabeledLoader
        self.UnlabeledLoader = UnlabeledLoader
        self.ValLoader = ValLoader
        self.TestLoader = TestLoader
        self.LabelList = LabelList
        self.NumLabels = len(LabelList)
        self.Epochs = Epochs
        self.PretrainEpochs = PretrainEpochs
        self.Lr = Lr
        self.ZetaId = ZetaId
        self.ZetaOod = ZetaOod
        self.XiOod = XiOod
        self.Wp = Wp
        self.Ws = Ws
        self.We = We
        self.Ww = Ww
        self.EmaMomentum = EmaMomentum
        self.Device = Device
        self.RunName = RunName
        self.TauId = None
        self.TauOod = None
        self.MarginOod = None
        self.EmaModel = None
    def free_energy(self, logits):
        return -torch.logsumexp(logits, dim=1)
    def update_ema(self):
        for ema_param, param in zip(self.EmaModel.parameters(), self.Model.parameters()):
            ema_param.data = self.EmaMomentum * ema_param.data + (1 - self.EmaMomentum) * param.data
    def compute_adaptive_thresholds(self):
        self.Model.eval()
        self.EmaModel.eval()
        energy_scores = []
        with torch.no_grad():
            for batch in tqdm(self.LabeledLoader, desc="Computing Thresholds", leave=False):
                input_ids = batch["OriginalInputIds"].to(self.Device)
                attention_mask = batch["OriginalAttentionMask"].to(self.Device)
                logits = self.EmaModel(input_ids, attention_mask)
                fe = self.free_energy(logits)
                energy_scores.extend(fe.cpu().numpy())
        energy_scores = np.array(energy_scores)
        S_m = np.median(energy_scores)
        S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        self.TauId = S_m - S_iqr * self.ZetaId
        self.TauOod = S_m + S_iqr * self.ZetaOod
        self.MarginOod = S_m + S_iqr * self.XiOod
        print(f"Adaptive thresholds calculated:")
        print(f"  Median energy (S_m): {S_m:.4f}")
        print(f"  IQR (S_iqr): {S_iqr:.4f}")
        print(f"  τ_id: {self.TauId:.4f}")
        print(f"  τ_ood: {self.TauOod:.4f}")
        print(f"  margin_ood: {self.MarginOod:.4f}")
        print(f"  Threshold gap: τ_ood - τ_id = {self.TauOod - self.TauId:.4f}")
        print(f"  Margin gap: m_ood - τ_ood = {self.MarginOod - self.TauOod:.4f}")
        self.Model.train()
    def Run(self):
        Optimizer = optim.AdamW(self.Model.parameters(), lr=self.Lr)
        TotalSteps = self.Epochs * len(self.LabeledLoader)
        Scheduler = get_cosine_schedule_with_warmup(
            Optimizer, num_warmup_steps=0, num_training_steps=TotalSteps
        )
        scaler = GradScaler()
        TrainLossList = []
        TrainSupLossList = []
        TrainSelfSupLossList = []
        TrainPseudoLossList = []
        TrainEnergyLossList = []
        TrainAccList = []
        ValLossList = []
        ValAccList = []
        # Pre-training
        self.EmaModel = deepcopy(self.Model).to(self.Device)
        for Ep in range(self.PretrainEpochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} Pretrain E{Ep+1}/{self.PretrainEpochs}",
                leave=False
            ):
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    WeakProjNorm = F.normalize(WeakProj, dim=1)
                    StrongProjNorm = F.normalize(StrongProj, dim=1)
                    LossSelfSup = -torch.mean(torch.sum(WeakProjNorm * StrongProjNorm, dim=1))
                Loss = LossSup + self.Ws * LossSelfSup
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                NumBatches += 1
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(0.0)
            TrainEnergyLossList.append(0.0)
            TrainAccList.append(TrainAcc)
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValEnergyScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} Pretrain E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValEnergyScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Pretrain Epoch[{Ep+1}/{self.PretrainEpochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val Energy mean: {np.mean(ValEnergyScores):.4f}, std: {np.std(ValEnergyScores):.4f}")
        # Compute thresholds
        self.compute_adaptive_thresholds()
        # Main training
        for Ep in range(self.PretrainEpochs, self.Epochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            PseudoLossTotal = 0.0
            EnergyLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} E{Ep+1}/{self.Epochs}",
                leave=False
            ):
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                IsIdBatch = UnlabeledBatch["IsId"].to(self.Device)
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    WeakProjNorm = F.normalize(WeakProj, dim=1)
                    StrongProjNorm = F.normalize(StrongProj, dim=1)
                    LossSelfSup = -torch.mean(torch.sum(WeakProjNorm * StrongProjNorm, dim=1))
                # Pseudo-labeling
                with torch.no_grad():
                    LogitsUWeak = self.Model(UWeakIds, UWeakMask)
                    FeWeak = self.free_energy(LogitsUWeak)
                    PseudoMask = (FeWeak < self.TauId).float() * IsIdBatch.float()
                    _, PseudoLabels = torch.max(F.softmax(LogitsUWeak, dim=1), dim=1)
                with autocast(device_type="cuda"):
                    LogitsUStrong = self.Model(UStrongIds, UStrongMask)
                    LossPseudo = F.cross_entropy(LogitsUStrong, PseudoLabels, reduction="none")
                    LossPseudo = (LossPseudo * PseudoMask).mean()
                    if PseudoMask.sum() == 0:
                        LossPseudo = torch.tensor(0.0, device=self.Device)
                # Energy regularization
                OutlierMask = (FeWeak > self.TauOod).float()
                if OutlierMask.sum() > 0:
                    LossEnergy = torch.mean(
                        OutlierMask * torch.relu(self.MarginOod - FeWeak) ** 2
                    ) / (OutlierMask.sum() + 1e-8)
                else:
                    LossEnergy = torch.tensor(0.0, device=self.Device)
                WeightNorm = sum(p.norm().pow(2) for p in self.Model.parameters()) * 0.5
                Loss = (
                    LossSup
                    + self.Wp * LossPseudo
                    + self.Ws * LossSelfSup
                    + self.We * LossEnergy
                    + self.Ww * WeightNorm
                )
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                Scheduler.step()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                PseudoLossTotal += LossPseudo.item()
                EnergyLossTotal += LossEnergy.item()
                NumBatches += 1
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            AvgPseudoLoss = PseudoLossTotal / NumBatches
            AvgEnergyLoss = EnergyLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(AvgPseudoLoss)
            TrainEnergyLossList.append(AvgEnergyLoss)
            TrainAccList.append(TrainAcc)
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValEnergyScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValEnergyScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Epoch[{Ep+1}/{self.Epochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f},")
            print(f"Pseudo={AvgPseudoLoss:.4f}, Energy={AvgEnergyLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val Energy mean: {np.mean(ValEnergyScores):.4f}, std: {np.std(ValEnergyScores):.4f}")
        TestPreds = []
        TestLabelsList = []
        TestProbsList = []
        TestEnergyScores = []
        self.Model.eval()
        with torch.no_grad():
            for Batch in tqdm(
                self.TestLoader,
                desc=f"{self.RunName} Test",
                leave=False
            ):
                Ids_ = Batch["InputIds"].to(self.Device)
                Mask_ = Batch["AttentionMask"].to(self.Device)
                Labels_ = Batch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    Logits_ = self.Model(Ids_, Mask_)
                    Probs_ = F.softmax(Logits_, dim=1)
                    Fe_ = self.free_energy(Logits_)
                Preds_ = torch.argmax(Logits_, dim=1)
                TestPreds.extend(Preds_.cpu().numpy())
                TestLabelsList.extend(Labels_.cpu().numpy())
                TestProbsList.extend(Probs_.cpu().numpy())
                TestEnergyScores.extend(Fe_.cpu().numpy())
        TestAcc = accuracy_score(TestLabelsList, TestPreds)
        OneHotLabels = np.eye(self.NumLabels)[np.array(TestLabelsList)]
        RocAuc = roc_auc_score(OneHotLabels, np.array(TestProbsList), multi_class="ovr")
        AvgPrecision = average_precision_score(OneHotLabels, np.array(TestProbsList), average="macro")
        print(f"\n{self.RunName} Final Test Results:")
        print(f"Test Accuracy: {TestAcc*100:.2f}%")
        print(f"ROC AUC: {RocAuc:.4f}, Average Precision: {AvgPrecision:.4f}")
        print(f"Test Energy mean: {np.mean(TestEnergyScores):.4f}, std: {np.std(TestEnergyScores):.4f}")
        print("Classification Report:")
        print(classification_report(TestLabelsList, TestPreds, target_names=self.LabelList, zero_division=0))
        return {
            "TrainLoss": TrainLossList,
            "TrainSupLoss": TrainSupLossList,
            "TrainSelfSupLoss": TrainSelfSupLossList,
            "TrainPseudoLoss": TrainPseudoLossList,
            "TrainEnergyLoss": TrainEnergyLossList,
            "TrainAcc": TrainAccList,
            "ValLoss": ValLossList,
            "ValAcc": ValAccList,
            "TestAcc": TestAcc,
            "RocAuc": RocAuc,
            "AvgPrecision": AvgPrecision
        }

def Main():
    print("Starting the training")
    nltk.download("wordnet")
    FilePath = r"/kaggle/input/pubmed-rct/pubmed_rct.txt"
    Loader = LoadPubmedRct(FilePath)
    DF = Loader.GetDataFrame()
    Splitter = SplitEightyTenTen(DF)
    TrainDF, ValDF, TestDF = Splitter.Split()
    Tokenizer = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
    IdCategories = ["METHODS", "RESULTS"]
    LabeledSizes = [500, 1000, 5000]
    print(f"ID Categories: {IdCategories}")
    ResultsDict = {}
    for Size in LabeledSizes:
        print(f"\nRunning SeFOSS With {Size} Labeled ID Samples")
        TrainIdDF = TrainDF[TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        TrainOodDF = TrainDF[~TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        if len(TrainIdDF) < Size:
            raise ValueError(f"Requested {Size} labeled samples, but only {len(TrainIdDF)} ID samples available.")
        LabeledSplit, UnlabeledIdSplit = train_test_split(
            TrainIdDF,
            train_size=Size,
            stratify=TrainIdDF["label"],
            random_state=RandomState
        )
        UnlabeledSplit = pd.concat([UnlabeledIdSplit, TrainOodDF]).reset_index(drop=True)
        print(f"Labeled ID: {len(LabeledSplit)}, Unlabeled: {len(UnlabeledSplit)} (ID={len(UnlabeledIdSplit)}, OOD={len(TrainOodDF)})")
        LabeledDataset = PubmedRctSeFOSSDataset(
            LabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False
        )
        UnlabeledDataset = PubmedRctSeFOSSDataset(
            UnlabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=True
        )
        ValData = PubmedRctTestDataset(ValDF, Tokenizer, IdCategories, MaxLen=256)
        TestData = PubmedRctTestDataset(TestDF, Tokenizer, IdCategories, MaxLen=256)
        LabeledLoader = DataLoader(LabeledDataset, batch_size=4, shuffle=True)
        UnlabeledLoader = DataLoader(UnlabeledDataset, batch_size=4, shuffle=True)
        ValLoader = DataLoader(ValData, batch_size=4, shuffle=False)
        TestLoader = DataLoader(TestData, batch_size=4, shuffle=False)
        Model = SeFOSS(num_classes=len(IdCategories), proj_dim=128)
        Trainer = TrainSeFOSSText(
            Model,
            LabeledLoader,
            UnlabeledLoader,
            ValLoader,
            TestLoader,
            IdCategories,
            Epochs=10,
            PretrainEpochs=1,
            Lr=2e-5,
            ZetaId=0.2,
            ZetaOod=1.5,
            XiOod=2.1,
            Wp=1.0,
            Ws=0.5,
            We=1.0,
            Ww=1e-4,
            EmaMomentum=0.999,
            Device=DEVICE,
            RunName=f"PubMedRCT_{Size}_Labeled"
        )
        RunResults = Trainer.Run()
        torch.save(Model.state_dict(), f"SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        print(f"Model weights saved to SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        EpochsRange = range(1, len(RunResults["TrainLoss"]) + 1)
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Total Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSupLoss"], label="Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSelfSupLoss"], label="Self-Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainPseudoLoss"], label="Pseudo-Labeling Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainEnergyLoss"], label="Energy Loss", marker="o")
        plt.title(f"Training Loss Curves ({Size} Labeled)")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainAcc"], label="Train Accuracy", marker="o")
        plt.title(f"Training Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Validation Loss", marker="o")
        plt.title(f"Validation Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, [v * 100 for v in RunResults["ValAcc"]], label="Val Accuracy (%)", marker="o")
        plt.title(f"Validation Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy (%)")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Val Loss", marker="o")
        plt.title(f"Train Vs. Val Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        ResultsDict[Size] = RunResults
        print(f"--- Finished Plotting For {Size} Labeled Samples ---")
    BestCount = None
    BestAcc = 0.0
    for Size in LabeledSizes:
        Acc = ResultsDict[Size]["TestAcc"]
        if Acc > BestAcc:
            BestAcc = Acc
            BestCount = Size
    print(f"\nBest Run: {BestCount} Labeled Samples With Test Accuracy: {BestAcc * 100:.2f}%")

if __name__ == "__main__":
    Main()
```

```python
import os
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from torch.amp import autocast, GradScaler
from transformers import AutoModel, AutoTokenizer, get_cosine_schedule_with_warmup
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, average_precision_score
from tqdm import tqdm
import nltk
from nltk.corpus import wordnet
import warnings
from copy import deepcopy

# Suppress CUDA warnings
warnings.filterwarnings("ignore", message="Unable to register cuFFT factory")
warnings.filterwarnings("ignore", message="Unable to register cuDNN factory")
warnings.filterwarnings("ignore", message="Unable to register cuBLAS factory")

# Set random seeds
RandomState = 2022469
np.random.seed(RandomState)
random.seed(RandomState)
torch.manual_seed(RandomState)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True

def cosine_similarity_loss(f1, f2):
    """
    Compute cosine similarity loss between two feature tensors (negative mean cosine similarity).
    Inputs are assumed to be projected features, as per SeFOSS paper Equation 2.
    """
    f1_norm = F.normalize(f1, p=2, dim=1)
    f2_norm = F.normalize(f2, p=2, dim=1)
    loss = - (f1_norm * f2_norm).sum(dim=1).mean()
    return loss

def LoadPubmedRctFile(FilePath):
    Cases = []
    with open(FilePath, "r", encoding="utf-8") as Fobj:
        Case = {"label": None, "text": ""}
        for Line in Fobj:
            Line = Line.strip()
            if Line == "":
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Case = {"label": None, "text": ""}
            elif Line.split("\t")[0].isupper():
                if Case["label"] and Case["text"]:
                    Cases.append(Case)
                Parts = Line.split("\t")
                Case = {"label": Parts[0], "text": Parts[1]}
            else:
                Case["text"] += " " + Line.strip()
        if Case["label"] and Case["text"]:
            Cases.append(Case)
    DF = pd.DataFrame(Cases)
    print(f"Dataset loaded with {len(DF)} samples.")
    print(DF.head())
    return DF

class LoadPubmedRct:
    def __init__(self, FilePath):
        self.DataFrame = LoadPubmedRctFile(FilePath)
    def GetDataFrame(self):
        return self.DataFrame

class SplitEightyTenTen:
    def __init__(self, DataFrame):
        self.DataFrame = DataFrame
    def Split(self):
        TrainData, TempData = train_test_split(
            self.DataFrame,
            test_size=0.2,
            stratify=self.DataFrame["label"],
            random_state=RandomState
        )
        ValData, TestData = train_test_split(
            TempData,
            test_size=0.5,
            stratify=TempData["label"],
            random_state=RandomState
        )
        print("Train:", len(TrainData), "Val:", len(ValData), "Test:", len(TestData))
        return TrainData, ValData, TestData

class AugSynonymReplace:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                Syns = wordnet.synsets(W)
                if Syns:
                    Lemmas = Syns[0].lemmas()
                    if Lemmas:
                        Synonym = Lemmas[0].name()
                        if Synonym != W:
                            NewWords.append(Synonym)
                            continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomMask:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
    def __call__(self, Text):
        Tokens = self.Tokenizer.tokenize(Text)
        if len(Tokens) == 0:
            return Text
        Idx = random.randint(0, len(Tokens)-1)
        Tokens[Idx] = self.Tokenizer.mask_token
        return self.Tokenizer.convert_tokens_to_string(Tokens)

class AugRandomDelete:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        NewWords = []
        for W in Words:
            if random.random() < self.Probability:
                continue
            NewWords.append(W)
        return " ".join(NewWords)

class AugRandomSwap:
    def __init__(self, Probability=0.15):
        self.Probability = Probability
    def __call__(self, Text):
        Words = Text.split()
        if len(Words) < 2:
            return Text
        if random.random() < self.Probability:
            i = random.randint(0, len(Words)-1)
            j = random.randint(0, len(Words)-1)
            while j == i:
                j = random.randint(0, len(Words)-1)
            Words[i], Words[j] = Words[j], Words[i]
        return " ".join(Words)

class WeakAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.1)
        self.Mask = AugRandomMask(Tokenizer)
    def __call__(self, Text):
        SynText = self.SynReplace(Text)
        MaskedText = self.Mask(SynText)
        return MaskedText

class StrongAugmentText:
    def __init__(self, Tokenizer):
        self.Tokenizer = Tokenizer
        self.SynReplace = AugSynonymReplace(Probability=0.2)
        self.Delete = AugRandomDelete(Probability=0.15)
        self.Swap = AugRandomSwap(Probability=0.15)
    def __call__(self, Text):
        SynText = self.SynReplace(Text)
        DeletedText = self.Delete(SynText)
        SwappedText = self.Swap(DeletedText)
        return SwappedText

class PubmedRctSeFOSSDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories
        self.MaxLen = MaxLen
        self.IsUnlabeled = IsUnlabeled
        self.LabelList = sorted(self.Df["label"].unique())
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.WeakAug = WeakAugmentText(Tokenizer)
        self.StrongAug = StrongAugmentText(Tokenizer)
        self.IsId = self.Df["label"].isin(self.IdCategories).values
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        IsIdSample = self.IsId[Idx]
        Label = self.Label2Idx[LabelStr] if IsIdSample or not self.IsUnlabeled else -1
        OriginalEnc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        WeakText = self.WeakAug(Text)
        WeakEnc = self.Tokenizer.encode_plus(
            WeakText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        StrongText = self.StrongAug(Text)
        StrongEnc = self.Tokenizer.encode_plus(
            StrongText,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "OriginalInputIds": OriginalEnc["input_ids"].squeeze(0),
            "OriginalAttentionMask": OriginalEnc["attention_mask"].squeeze(0),
            "WeakInputIds": WeakEnc["input_ids"].squeeze(0),
            "WeakAttentionMask": WeakEnc["attention_mask"].squeeze(0),
            "StrongInputIds": StrongEnc["input_ids"].squeeze(0),
            "StrongAttentionMask": StrongEnc["attention_mask"].squeeze(0),
            "Label": Label,
            "IsId": IsIdSample
        }

class PubmedRctTestDataset(Dataset):
    def __init__(self, DataFrame, Tokenizer, IdCategories, MaxLen=256):
        super().__init__()
        self.Df = DataFrame.reset_index(drop=True)
        self.Tokenizer = Tokenizer
        self.IdCategories = IdCategories
        self.MaxLen = MaxLen
        self.LabelList = sorted(set(IdCategories))
        self.Label2Idx = {Lab: i for i, Lab in enumerate(self.LabelList)}
        self.Df = self.Df[self.Df["label"].isin(self.IdCategories)].reset_index(drop=True)
    def __len__(self):
        return len(self.Df)
    def __getitem__(self, Idx):
        Row = self.Df.iloc[Idx]
        Text = Row["text"]
        LabelStr = Row["label"]
        Label = self.Label2Idx[LabelStr]
        Enc = self.Tokenizer.encode_plus(
            Text,
            truncation=True,
            max_length=self.MaxLen,
            return_tensors="pt",
            padding="max_length"
        )
        return {
            "InputIds": Enc["input_ids"].squeeze(0),
            "AttentionMask": Enc["attention_mask"].squeeze(0),
            "Label": Label
        }

class SeFOSS(nn.Module):
    def __init__(self, num_classes, proj_dim=128):
        super(SeFOSS, self).__init__()
        self.backbone = AutoModel.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
        self.feature_dim = self.backbone.config.hidden_size
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)
    def forward(self, input_ids, attention_mask):
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        features = outputs.last_hidden_state[:, 0]
        logits = self.classifier(features)
        return logits
    def forward_features(self, input_ids, attention_mask):
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        return outputs.last_hidden_state[:, 0]
    def forward_projection(self, features):
        return self.projection(features)

class TrainSeFOSSText:
    def __init__(
        self,
        Model,
        LabeledLoader,
        UnlabeledLoader,
        ValLoader,
        TestLoader,
        LabelList,
        Epochs=10,
        PretrainEpochs=1,
        Lr=2e-5,
        ZetaId=0.2,
        ZetaOod=1.5,
        XiOod=2.1,
        Wp=1.0,
        Ws=0.5,
        We=1.0,
        Ww=1e-4,
        EmaMomentum=0.999,
        Device=DEVICE,
        RunName="Run"
    ):
        self.Model = Model.to(Device)
        self.LabeledLoader = LabeledLoader
        self.UnlabeledLoader = UnlabeledLoader
        self.ValLoader = ValLoader
        self.TestLoader = TestLoader
        self.LabelList = LabelList
        self.NumLabels = len(LabelList)
        self.Epochs = Epochs
        self.PretrainEpochs = PretrainEpochs
        self.Lr = Lr
        self.ZetaId = ZetaId
        self.ZetaOod = ZetaOod
        self.XiOod = XiOod
        self.Wp = Wp
        self.Ws = Ws
        self.We = We
        self.Ww = Ww
        self.EmaMomentum = EmaMomentum
        self.Device = Device
        self.RunName = RunName
        self.TauId = None
        self.TauOod = None
        self.MarginOod = None
        self.EmaModel = None
    def free_energy(self, logits):
        return -torch.logsumexp(logits, dim=1)
    def update_ema(self):
        for ema_param, param in zip(self.EmaModel.parameters(), self.Model.parameters()):
            ema_param.data = self.EmaMomentum * ema_param.data + (1 - self.EmaMomentum) * param.data
    def compute_adaptive_thresholds(self):
        self.Model.eval()
        self.EmaModel.eval()
        energy_scores = []
        with torch.no_grad():
            for batch in tqdm(self.LabeledLoader, desc="Computing Thresholds", leave=False):
                input_ids = batch["OriginalInputIds"].to(self.Device)
                attention_mask = batch["OriginalAttentionMask"].to(self.Device)
                logits = self.EmaModel(input_ids, attention_mask)
                fe = self.free_energy(logits)
                energy_scores.extend(fe.cpu().numpy())
        energy_scores = np.array(energy_scores)
        S_m = np.median(energy_scores)
        S_iqr = np.percentile(energy_scores, 75) - np.percentile(energy_scores, 25)
        self.TauId = S_m - S_iqr * self.ZetaId
        self.TauOod = S_m + S_iqr * self.ZetaOod
        self.MarginOod = S_m + S_iqr * self.XiOod
        print(f"Adaptive thresholds calculated:")
        print(f"  Median energy (S_m): {S_m:.4f}")
        print(f"  IQR (S_iqr): {S_iqr:.4f}")
        print(f"  τ_id: {self.TauId:.4f}")
        print(f"  τ_ood: {self.TauOod:.4f}")
        print(f"  margin_ood: {self.MarginOod:.4f}")
        print(f"  Threshold gap: τ_ood - τ_id = {self.TauOod - self.TauId:.4f}")
        print(f"  Margin gap: m_ood - τ_ood = {self.MarginOod - self.TauOod:.4f}")
        self.Model.train()
    def Run(self):
        Optimizer = optim.AdamW(self.Model.parameters(), lr=self.Lr)
        TotalSteps = self.Epochs * len(self.LabeledLoader)
        Scheduler = get_cosine_schedule_with_warmup(
            Optimizer, num_warmup_steps=0, num_training_steps=TotalSteps
        )
        scaler = GradScaler()
        TrainLossList = []
        TrainSupLossList = []
        TrainSelfSupLossList = []
        TrainPseudoLossList = []
        TrainEnergyLossList = []
        TrainAccList = []
        ValLossList = []
        ValAccList = []
        # Pre-training
        self.EmaModel = deepcopy(self.Model).to(self.Device)
        for Ep in range(self.PretrainEpochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} Pretrain E{Ep+1}/{self.PretrainEpochs}",
                leave=False
            ):
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    LossSelfSup = cosine_similarity_loss(WeakProj, StrongProj)
                Loss = LossSup + self.Ws * LossSelfSup
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                NumBatches += 1
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(0.0)
            TrainEnergyLossList.append(0.0)
            TrainAccList.append(TrainAcc)
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValEnergyScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} Pretrain E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValEnergyScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Pretrain Epoch[{Ep+1}/{self.PretrainEpochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val Energy mean: {np.mean(ValEnergyScores):.4f}, std: {np.std(ValEnergyScores):.4f}")
        # Compute thresholds
        self.compute_adaptive_thresholds()
        # Main training
        for Ep in range(self.PretrainEpochs, self.Epochs):
            self.Model.train()
            TotalLoss = 0.0
            SupLossTotal = 0.0
            SelfSupLossTotal = 0.0
            PseudoLossTotal = 0.0
            EnergyLossTotal = 0.0
            CorrectSup = 0
            TotalSup = 0
            NumBatches = 0
            UnlabeledIter = iter(self.UnlabeledLoader)
            for LabeledBatch in tqdm(
                self.LabeledLoader,
                desc=f"{self.RunName} E{Ep+1}/{self.Epochs}",
                leave=False
            ):
                InputsWeakIds = LabeledBatch["WeakInputIds"].to(self.Device)
                InputsWeakMask = LabeledBatch["WeakAttentionMask"].to(self.Device)
                LabelsL = LabeledBatch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    LogitsL = self.Model(InputsWeakIds, InputsWeakMask)
                    LossSup = F.cross_entropy(LogitsL, LabelsL)
                PredsL = torch.argmax(LogitsL, dim=1)
                CorrectSup += (PredsL == LabelsL).sum().item()
                TotalSup += LabelsL.size(0)
                try:
                    UnlabeledBatch = next(UnlabeledIter)
                except StopIteration:
                    UnlabeledIter = iter(self.UnlabeledLoader)
                    UnlabeledBatch = next(UnlabeledIter)
                UWeakIds = UnlabeledBatch["WeakInputIds"].to(self.Device)
                UWeakMask = UnlabeledBatch["WeakAttentionMask"].to(self.Device)
                UStrongIds = UnlabeledBatch["StrongInputIds"].to(self.Device)
                UStrongMask = UnlabeledBatch["StrongAttentionMask"].to(self.Device)
                IsIdBatch = UnlabeledBatch["IsId"].to(self.Device)
                with autocast(device_type="cuda"):
                    WeakFeatures = self.Model.forward_features(UWeakIds, UWeakMask)
                    StrongFeatures = self.Model.forward_features(UStrongIds, UStrongMask)
                    WeakProj = self.Model.forward_projection(WeakFeatures)
                    StrongProj = self.Model.forward_projection(StrongFeatures)
                    LossSelfSup = cosine_similarity_loss(WeakProj, StrongProj)
                # Pseudo-labeling
                with torch.no_grad():
                    LogitsUWeak = self.Model(UWeakIds, UWeakMask)
                    FeWeak = self.free_energy(LogitsUWeak)
                    PseudoMask = (FeWeak < self.TauId).float() * IsIdBatch.float()
                    _, PseudoLabels = torch.max(F.softmax(LogitsUWeak, dim=1), dim=1)
                with autocast(device_type="cuda"):
                    LogitsUStrong = self.Model(UStrongIds, UStrongMask)
                    LossPseudo = F.cross_entropy(LogitsUStrong, PseudoLabels, reduction="none")
                    LossPseudo = (LossPseudo * PseudoMask).mean()
                    if PseudoMask.sum() == 0:
                        LossPseudo = torch.tensor(0.0, device=self.Device)
                # Energy regularization
                OutlierMask = (FeWeak > self.TauOod).float()
                if OutlierMask.sum() > 0:
                    LossEnergy = torch.mean(
                        OutlierMask * torch.relu(self.MarginOod - FeWeak) ** 2
                    ) / (OutlierMask.sum() + 1e-8)
                else:
                    LossEnergy = torch.tensor(0.0, device=self.Device)
                WeightNorm = sum(p.norm().pow(2) for p in self.Model.parameters()) * 0.5
                Loss = (
                    LossSup
                    + self.Wp * LossPseudo
                    + self.Ws * LossSelfSup
                    + self.We * LossEnergy
                    + self.Ww * WeightNorm
                )
                Optimizer.zero_grad()
                scaler.scale(Loss).backward()
                torch.nn.utils.clip_grad_norm_(self.Model.parameters(), 1.0)
                scaler.step(Optimizer)
                scaler.update()
                self.update_ema()
                Scheduler.step()
                TotalLoss += Loss.item()
                SupLossTotal += LossSup.item()
                SelfSupLossTotal += LossSelfSup.item()
                PseudoLossTotal += LossPseudo.item()
                EnergyLossTotal += LossEnergy.item()
                NumBatches += 1
            AvgLoss = TotalLoss / NumBatches
            AvgSupLoss = SupLossTotal / NumBatches
            AvgSelfSupLoss = SelfSupLossTotal / NumBatches
            AvgPseudoLoss = PseudoLossTotal / NumBatches
            AvgEnergyLoss = EnergyLossTotal / NumBatches
            TrainAcc = CorrectSup / TotalSup
            TrainLossList.append(AvgLoss)
            TrainSupLossList.append(AvgSupLoss)
            TrainSelfSupLossList.append(AvgSelfSupLoss)
            TrainPseudoLossList.append(AvgPseudoLoss)
            TrainEnergyLossList.append(AvgEnergyLoss)
            TrainAccList.append(TrainAcc)
            self.Model.eval()
            ValLossVal = 0.0
            ValBatches = 0
            ValPreds = []
            ValLabelsList = []
            ValEnergyScores = []
            with torch.no_grad():
                for Batch in tqdm(
                    self.ValLoader,
                    desc=f"{self.RunName} E{Ep+1} Val",
                    leave=False
                ):
                    Ids_ = Batch["InputIds"].to(self.Device)
                    Mask_ = Batch["AttentionMask"].to(self.Device)
                    Labels_ = Batch["Label"].to(self.Device)
                    with autocast(device_type="cuda"):
                        Logits_ = self.Model(Ids_, Mask_)
                        Loss_ = F.cross_entropy(Logits_, Labels_)
                        Fe_ = self.free_energy(Logits_)
                    ValLossVal += Loss_.item()
                    ValBatches += 1
                    Preds_ = torch.argmax(Logits_, dim=1)
                    ValPreds.extend(Preds_.cpu().numpy())
                    ValLabelsList.extend(Labels_.cpu().numpy())
                    ValEnergyScores.extend(Fe_.cpu().numpy())
            AvgValLoss = ValLossVal / ValBatches
            ValAccuracy = accuracy_score(ValLabelsList, ValPreds)
            ValLossList.append(AvgValLoss)
            ValAccList.append(ValAccuracy)
            print(f"{self.RunName} Epoch[{Ep+1}/{self.Epochs}]")
            print(f"TrainLoss: {AvgLoss:.4f} (Sup={AvgSupLoss:.4f}, SelfSup={AvgSelfSupLoss:.4f},")
            print(f"Pseudo={AvgPseudoLoss:.4f}, Energy={AvgEnergyLoss:.4f})")
            print(f"TrainAcc: {TrainAcc*100:.2f}%")
            print(f"ValLoss: {AvgValLoss:.4f} ValAcc: {ValAccuracy*100:.2f}%")
            print(f"Val Energy mean: {np.mean(ValEnergyScores):.4f}, std: {np.std(ValEnergyScores):.4f}")
        TestPreds = []
        TestLabelsList = []
        TestProbsList = []
        TestEnergyScores = []
        self.Model.eval()
        with torch.no_grad():
            for Batch in tqdm(
                self.TestLoader,
                desc=f"{self.RunName} Test",
                leave=False
            ):
                Ids_ = Batch["InputIds"].to(self.Device)
                Mask_ = Batch["AttentionMask"].to(self.Device)
                Labels_ = Batch["Label"].to(self.Device)
                with autocast(device_type="cuda"):
                    Logits_ = self.Model(Ids_, Mask_)
                    Probs_ = F.softmax(Logits_, dim=1)
                    Fe_ = self.free_energy(Logits_)
                Preds_ = torch.argmax(Logits_, dim=1)
                TestPreds.extend(Preds_.cpu().numpy())
                TestLabelsList.extend(Labels_.cpu().numpy())
                TestProbsList.extend(Probs_.cpu().numpy())
                TestEnergyScores.extend(Fe_.cpu().numpy())
        TestAcc = accuracy_score(TestLabelsList, TestPreds)
        OneHotLabels = np.eye(self.NumLabels)[np.array(TestLabelsList)]
        RocAuc = roc_auc_score(OneHotLabels, np.array(TestProbsList), multi_class="ovr")
        AvgPrecision = average_precision_score(OneHotLabels, np.array(TestProbsList), average="macro")
        print(f"\n{self.RunName} Final Test Results:")
        print(f"Test Accuracy: {TestAcc*100:.2f}%")
        print(f"ROC AUC: {RocAuc:.4f}, Average Precision: {AvgPrecision:.4f}")
        print(f"Test Energy mean: {np.mean(TestEnergyScores):.4f}, std: {np.std(TestEnergyScores):.4f}")
        print("Classification Report:")
        print(classification_report(TestLabelsList, TestPreds, target_names=self.LabelList, zero_division=0))
        return {
            "TrainLoss": TrainLossList,
            "TrainSupLoss": TrainSupLossList,
            "TrainSelfSupLoss": TrainSelfSupLossList,
            "TrainPseudoLoss": TrainPseudoLossList,
            "TrainEnergyLoss": TrainEnergyLossList,
            "TrainAcc": TrainAccList,
            "ValLoss": ValLossList,
            "ValAcc": ValAccList,
            "TestAcc": TestAcc,
            "RocAuc": RocAuc,
            "AvgPrecision": AvgPrecision
        }

def Main():
    print("Starting the training")
    nltk.download("wordnet")
    FilePath = r"/kaggle/input/pubmed-rct/pubmed_rct.txt"
    Loader = LoadPubmedRct(FilePath)
    DF = Loader.GetDataFrame()
    Splitter = SplitEightyTenTen(DF)
    TrainDF, ValDF, TestDF = Splitter.Split()
    Tokenizer = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
    IdCategories = ["METHODS", "RESULTS"]
    LabeledSizes = [500, 1000, 5000]
    print(f"ID Categories: {IdCategories}")
    ResultsDict = {}
    for Size in LabeledSizes:
        print(f"\nRunning SeFOSS With {Size} Labeled ID Samples")
        TrainIdDF = TrainDF[TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        TrainOodDF = TrainDF[~TrainDF["label"].isin(IdCategories)].reset_index(drop=True)
        if len(TrainIdDF) < Size:
            raise ValueError(f"Requested {Size} labeled samples, but only {len(TrainIdDF)} ID samples available.")
        LabeledSplit, UnlabeledIdSplit = train_test_split(
            TrainIdDF,
            train_size=Size,
            stratify=TrainIdDF["label"],
            random_state=RandomState
        )
        UnlabeledSplit = pd.concat([UnlabeledIdSplit, TrainOodDF]).reset_index(drop=True)
        print(f"Labeled ID: {len(LabeledSplit)}, Unlabeled: {len(UnlabeledSplit)} (ID={len(UnlabeledIdSplit)}, OOD={len(TrainOodDF)})")
        LabeledDataset = PubmedRctSeFOSSDataset(
            LabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=False
        )
        UnlabeledDataset = PubmedRctSeFOSSDataset(
            UnlabeledSplit, Tokenizer, IdCategories, MaxLen=256, IsUnlabeled=True
        )
        ValData = PubmedRctTestDataset(ValDF, Tokenizer, IdCategories, MaxLen=256)
        TestData = PubmedRctTestDataset(TestDF, Tokenizer, IdCategories, MaxLen=256)
        LabeledLoader = DataLoader(LabeledDataset, batch_size=4, shuffle=True)
        UnlabeledLoader = DataLoader(UnlabeledDataset, batch_size=4, shuffle=True)
        ValLoader = DataLoader(ValData, batch_size=4, shuffle=False)
        TestLoader = DataLoader(TestData, batch_size=4, shuffle=False)
        Model = SeFOSS(num_classes=len(IdCategories), proj_dim=128)
        Trainer = TrainSeFOSSText(
            Model,
            LabeledLoader,
            UnlabeledLoader,
            ValLoader,
            TestLoader,
            IdCategories,
            Epochs=5,
            PretrainEpochs=1,
            Lr=2e-5,
            ZetaId=0.2,
            ZetaOod=1.5,
            XiOod=2.1,
            Wp=1.0,
            Ws=0.5,
            We=1.0,
            Ww=1e-4,
            EmaMomentum=0.999,
            Device=DEVICE,
            RunName=f"PubMedRCT_{Size}_Labeled"
        )
        RunResults = Trainer.Run()
        torch.save(Model.state_dict(), f"SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        print(f"Model weights saved to SeFOSS_PubMed_RCT_{Size}_labeled.pth")
        EpochsRange = range(1, len(RunResults["TrainLoss"]) + 1)
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Total Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSupLoss"], label="Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainSelfSupLoss"], label="Self-Supervised Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainPseudoLoss"], label="Pseudo-Labeling Loss", marker="o")
        plt.plot(EpochsRange, RunResults["TrainEnergyLoss"], label="Energy Loss", marker="o")
        plt.title(f"Training Loss Curves ({Size} Labeled)")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainAcc"], label="Train Accuracy", marker="o")
        plt.title(f"Training Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Validation Loss", marker="o")
        plt.title(f"Validation Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, [v * 100 for v in RunResults["ValAcc"]], label="Val Accuracy (%)", marker="o")
        plt.title(f"Validation Accuracy (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy (%)")
        plt.legend()
        plt.tight_layout()
        plt.show()
        plt.figure(figsize=(8, 5))
        plt.plot(EpochsRange, RunResults["TrainLoss"], label="Train Loss", marker="o")
        plt.plot(EpochsRange, RunResults["ValLoss"], label="Val Loss", marker="o")
        plt.title(f"Train Vs. Val Loss (Size={Size})")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.tight_layout()
        plt.show()
        ResultsDict[Size] = RunResults
        print(f"--- Finished Plotting For {Size} Labeled Samples ---")
    BestCount = None
    BestAcc = 0.0
    for Size in LabeledSizes:
        Acc = ResultsDict[Size]["TestAcc"]
        if Acc > BestAcc:
            BestAcc = Acc
            BestCount = Size
    print(f"\nBest Run: {BestCount} Labeled Samples With Test Accuracy: {BestAcc * 100:.2f}%")

if __name__ == "__main__":
    Main()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import AutoModel, AutoTokenizer
import numpy as np

# Device configuration: use GPU if available, otherwise CPU
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the SeFOSS model architecture (must match training configuration)
class SeFOSS(nn.Module):
    def _init_(self, num_classes, proj_dim=128):
        super(SeFOSS, self)._init_()
        self.backbone = AutoModel.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")
        self.feature_dim = self.backbone.config.hidden_size  # e.g., 768
        self.classifier = nn.Linear(self.feature_dim, num_classes)
        self.projection = nn.Linear(self.feature_dim, proj_dim)
        
    def forward(self, input_ids, attention_mask):
        outputs = self.backbone(input_ids, attention_mask=attention_mask)
        features = outputs.last_hidden_state[:, 0]  # CLS token representation
        logits = self.classifier(features)
        return logits

# Define label mapping for the 2-class model (METHODS and RESULTS)
LABELS = {
    "METHODS": 0,
    "RESULTS": 1
}
LABEL_LIST = ["METHODS", "RESULTS"]
NUM_LABELS = len(LABEL_LIST)

# Load the tokenizer (must match your model from training)
TOKENIZER = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract")

# Specify the path to your trained model weights (adjust this path as needed)
MODEL_PATH = r"C:\Users\clash\Downloads\mkc\SeFOSS_PubMed_RCT_1000_labeled.pth"  # update with your file path

# Initialize the model with 2 classes, load the saved weights, and set to evaluation mode
model = SeFOSS(num_classes=NUM_LABELS, proj_dim=128)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.to(DEVICE)
model.eval()

# Adaptive thresholds (as computed previously)
tau_id = -0.3455
tau_ood = 0.1156
margin_ood = 0.2783

def classify_text(text, max_length=256):
    """
    Tokenizes the input text, runs inference, and returns:
      - predicted label (if in distribution),
      - confidence (softmax probability),
      - free-energy score,
      - full probabilities,
      - a flag whether the sample is considered in-distribution.
    
    A free-energy score below tau_id is treated as in-distribution. If the free-energy score
    is above tau_ood, the sample is flagged as out-of-distribution (OOD).
    """
    # Tokenize the input text
    encodings = TOKENIZER.encode_plus(
        text,
        truncation=True,
        max_length=max_length,
        return_tensors="pt",
        padding="max_length"
    )
    input_ids = encodings["input_ids"].to(DEVICE)
    attention_mask = encodings["attention_mask"].to(DEVICE)
    
    with torch.no_grad():
        logits = model(input_ids, attention_mask)
        # Compute softmax probabilities
        probs = F.softmax(logits, dim=1).cpu().numpy()[0]
        # Compute free-energy score: -logsumexp(logits)
        free_energy = -torch.logsumexp(logits, dim=1).cpu().numpy()[0]
    
    predicted_idx = np.argmax(probs)
    predicted_label = LABEL_LIST[predicted_idx]
    confidence = probs[predicted_idx]
    
    # Determine if sample is in distribution or OOD based on free-energy thresholds
    if free_energy < tau_id:
        distribution_status = "In-Distribution (Confident)"
    elif free_energy > tau_ood:
        distribution_status = "Out-Of-Distribution (OOD)"
    else:
        distribution_status = "Uncertain Region"

    return predicted_label, confidence, free_energy, probs, distribution_status

# Set your text input here
text_input = ("To compare the differences in the efficacy on lower limb motor function in children with spastic cerebral palsy between the combined therapeutic program of Jin three-needle therapy and MOTOmed intelligent motor training and the simple MOTOmed .")

# Classify the text
predicted_label, confidence, free_energy, all_probs, distribution_status = classify_text(text_input)

print("Input text:")
print(text_input)
print("\nPredicted Label:", predicted_label)
print("Confidence: {:.2f}%".format(confidence * 100))
print("Free-Energy Score: {:.4f}".format(free_energy))
print("Distribution Status:", distribution_status)
print("\nFull probabilities for each label:")
for label, prob in zip(LABEL_LIST, all_probs):
    print("  {}: {:.2f}%".format(label, prob * 100))
```
